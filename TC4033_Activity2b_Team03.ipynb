### Download Cifar10 dataset

For this exercise, the `CIFAR10` dataset is used. This is a popular and widely used dataset for evaluating or training of deep learning models. It contains a large group of images with 10 classes for different animals and objects.

In this part of the code, the dataset for the notebook is downloaded, configured and split as needed.
In this case, a value of 50,000 samples is used for the Training set with a mini-batch size of 64. It is recommended to always use powers of 2 as the mini-batch size for better memory handling.

There's three main concepts to explain in this section of code:
* Dataset download: For each set of data (Train, Validation and Test), a `cifar10` dataset object is created based on the specified data path. This object will contain all x and y values for the dataset. By setting the argument `train` to `True` or `False` the dataset portion where data is gonna be retrieved is defined, either from the Training or the Test portion of `CIFAR 10`.
* Dataset transformation: The downloaded data, in this case for the cifar10 dataset, is transformed based on the `transform` argument. By using `T.Compose`, a function from the torchvision.transforms library in PyTorch that allows you to combine multiple transformations into one, 3 transformations are applied particularly to this case, including the resizing of the input images to 224x224 pixels. Also, the input is converted in tensor form and data is normalized. The values used for normalization (mean and standard deviation for each one of the 3 RGB channels of the image) are known values for the CIFAR10 dataset to achieve a standard deviation of 1 and a mean of 0.
* Dataset Dataloader: A Dataloader object is created out of the defined dataset. This is an iterable object which contains all the minibatches generated from the previously created (or like in this case, downloaded) dataset. The data is sorted randomly to avoid undesired biases during the training process and also the size of the sample is set based on the previously defined values.

In summary, this code prepares and loads the CIFAR-10 training, validation, and test datasets, applying specific transformations and managing the data in mini-batches.

It is important to mention that if the dataset has already been downloaded, the output of this section will just show a message acknowledging this.

```python
# Path to store/load the CIFAR-10 dataset
DATA_PATH = 'Data/'

# Dataset split sizes
NUM_TRAIN = 50000    # Number of training samples
NUM_VAL = 5000      # Number of validation samples
NUM_TEST = 5000     # Number of test samples

# Training hyperparameters
MINIBATCH_SIZE = 64  # Size of each training batch

# Define data transformations for CIFAR-10
# Compose combines multiple transforms into a single transform
transform_cifar = T.Compose([
    T.ToTensor(),  # Convert PIL image to tensor (0-1 range)
    # Normalize using CIFAR-10 mean and std for each channel (RGB)
    # Mean: [0.491, 0.482, 0.447], Std: [0.247, 0.243, 0.261]
    T.Normalize([0.491, 0.482, 0.447], [0.247, 0.243, 0.261])
])

# Initialize training dataset
print(f"Loading training dataset from {DATA_PATH}...")
cifar10_train = datasets.CIFAR10(
    DATA_PATH,
    train=True,
    download=True,
    transform=transform_cifar
)
train_loader = DataLoader(
    cifar10_train,
    batch_size=MINIBATCH_SIZE,
    sampler=sampler.SubsetRandomSampler(range(NUM_TRAIN))  # Randomly sample training data
)
print(f"Training loader initialized with {NUM_TRAIN} samples, batch size: {MINIBATCH_SIZE}")

# Initialize validation dataset
print("\nLoading validation dataset...")
cifar10_val = datasets.CIFAR10(
    DATA_PATH,
    train=False,
    download=True,
    transform=transform_cifar
)
val_loader = DataLoader(
    cifar10_val,
    batch_size=MINIBATCH_SIZE,
    sampler=sampler.SubsetRandomSampler(range(NUM_VAL))  # Sample first NUM_VAL images for validation
)
print(f"Validation loader initialized with {NUM_VAL} samples")

# Initialize test dataset
print("\nLoading test dataset...")
cifar10_test = datasets.CIFAR10(
    DATA_PATH,
    train=False,
    download=True,
    transform=transform_cifar
)
test_loader = DataLoader(
    cifar10_test,
    batch_size=MINIBATCH_SIZE,
    # Sample remaining images after validation set for testing
    sampler=sampler.SubsetRandomSampler(range(NUM_VAL, len(cifar10_test)))
)
print(f"Test loader initialized with {NUM_TEST} samples")

# Print summary of data loading
print("\nDataset Summary:")
print(f"Total samples: {NUM_TRAIN + NUM_VAL + NUM_TEST}")
print(f"Number of batches - Train: {len(train_loader)}, Val: {len(val_loader)}, Test: {len(test_loader)}")
```

```python
def explore_cifar10_dataset(train_loader, val_loader, test_loader, cifar10_train, n_samples=5):
    """
    Comprehensive exploration of CIFAR-10 dataset including visualizations and statistics
    """
    try:
        # CIFAR-10 classes
        classes = ('plane', 'car', 'bird', 'cat', 'deer',
                  'dog', 'frog', 'horse', 'ship', 'truck')

        print("\nüéØ CIFAR-10 Dataset Explorer")
        print("=" * 50)

        # ... (rest of the code before show_samples remains unchanged) ...
        # Print CIFAR-10 Training Dataset Details
        print("\nüìö CIFAR-10 Training Dataset Structure")
        print("=" * 50)
        print(f"cifar10_train:".center(50))
        print("=" * 50)
        print(cifar10_train)
        print("=" * 50)

        # Calculate training data statistics
        total_samples = len(cifar10_train)
        batch_size = train_loader.batch_size
        partial_batch = total_samples % batch_size != 0
        full_mini_batches = int(total_samples/batch_size)
        total_mini_batches = full_mini_batches + partial_batch

        # Get tensor shape from first batch
        first_batch = next(iter(train_loader))
        tensors_shape = first_batch[0]  # Get the first element (images) from the batch
        tensor_shape = torch.tensor(tensors_shape.shape)

        # Print Training Data Configuration
        print("\nüìä Training Data Configuration")
        print("=" * 50)
        print(f"train_loader:".center(50))
        print("=" * 50)
        print("\nüîç Detailed Statistics:")
        print(f" ‚îú‚îÄ Total samples: {total_samples:,}")
        print(f" ‚îú‚îÄ Batch Size: {batch_size}")
        print(f" ‚îú‚îÄ Tensor Shape: {tensor_shape}")
        print(f" ‚îú‚îÄ Total Mini Batches: {total_mini_batches:,}")
        print(f" ‚îú‚îÄ Full Mini Batches: {full_mini_batches:,} (size: {tensor_shape[0]})")
        if partial_batch:
            print(f" ‚îî‚îÄ Partial Mini Batch: 1")
        else:
            print(f" ‚îî‚îÄ No partial mini batch")

        # Dataset size information
        print("\nüìà Dataset Split Statistics")
        print("=" * 50)
        print(f"üî∏ Training samples: {len(train_loader.sampler):,}")
        print(f"üî∏ Validation samples: {len(val_loader.sampler):,}")
        print(f"üî∏ Test samples: {len(test_loader.sampler):,}")
        print(f"üî∏ Batch size: {train_loader.batch_size}")
        print(f"üî∏ Number of classes: {len(classes)}")

        # Memory usage estimation
        def estimate_memory(loader):
            try:
                batch = next(iter(loader))
                images, _ = batch
                return images.element_size() * images.nelement() * len(loader.sampler) / (1024 * 1024)
            except Exception as e:
                print(f"Warning: Could not estimate memory for loader: {str(e)}")
                return 0

        print("\nüíæ Memory Requirements")
        print("=" * 50)
        print(f"üî∏ Training set: {estimate_memory(train_loader):.2f} MB")
        print(f"üî∏ Validation set: {estimate_memory(val_loader):.2f} MB")
        print(f"üî∏ Test set: {estimate_memory(test_loader):.2f} MB")

        # Image statistics
        try:
            batch = next(iter(train_loader))
            images, _ = batch

            print("\nüìê Image Properties")
            print("=" * 50)
            print(f"üî∏ Image dimensions: {tuple(images[0].shape)}")
            print(f"üî∏ Value range: [{images.min():.2f}, {images.max():.2f}]")
            print(f"üî∏ Mean: {images.mean().item():.3f}")
            print(f"üî∏ Std: {images.std().item():.3f}")
        except Exception as e:
            print(f"Warning: Could not compute image statistics: {str(e)}")

        # Class distribution visualization
        def get_class_distribution(loader):
            try:
                labels = []
                for _, batch_labels in loader:
                    if batch_labels.is_cuda:
                        batch_labels = batch_labels.cpu()
                    labels.extend(batch_labels.numpy())
                return Counter(labels)
            except Exception as e:
                print(f"Warning: Could not compute class distribution: {str(e)}")
                return Counter()

        train_dist = get_class_distribution(train_loader)

        if train_dist:
            # Create figure with custom size
            plt.figure(figsize=(14, 7))

            # Custom color palette - modern and vibrant colors
            colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4',
                      '#FFEEAD', '#D4A5A5', '#9B59B6', '#3498DB',
                      '#FF9F1C', '#2ECC71']

            # Set the background color
            plt.rcParams['axes.facecolor'] = '#f8f9fa'
            plt.rcParams['figure.facecolor'] = 'white'

            # Create bar plot
            bars = plt.bar(classes,
                            [train_dist[i] for i in range(len(classes))],
                            color=colors,
                            width=0.6)

            # Customize plot
            plt.title("Class Distribution in Training Set",
                    fontsize=16,
                    pad=20,
                    fontweight='bold')

            plt.xlabel("Classes", fontsize=12, labelpad=10)
            plt.ylabel("Number of Samples", fontsize=12, labelpad=10)

            # Rotate x-labels for better readability
            plt.xticks(rotation=30, ha='right', fontsize=10)
            plt.yticks(fontsize=10)

            # Add value labels on top of each bar
            for bar in bars:
                height = bar.get_height()
                plt.text(bar.get_x() + bar.get_width()/2., height,
                        f'{int(height):,}',
                        ha='center', va='bottom',
                        fontweight='bold')

            # Customize grid
            plt.grid(True, axis='y', linestyle='--', alpha=0.7)

            # Remove top and right spines
            ax = plt.gca()
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)

            # Add some padding to the y-axis to make room for value labels
            plt.margins(y=0.1)

            # Adjust layout
            plt.tight_layout()

            # Show plot
            plt.show()


        # Sample images visualization (Modified)
        def show_samples(loader, classes, n_samples):
            try:
                class_images = {i: [] for i in range(len(classes))}
                for images, labels in loader:
                    if images.is_cuda:
                        images = images.cpu()
                    if labels.is_cuda:
                        labels = labels.cpu()
                    for img, label in zip(images, labels):
                        if len(class_images[label.item()]) < n_samples:
                            class_images[label.item()].append(img)
                    if all(len(imgs) >= n_samples for imgs in class_images.values()):
                        break

                # Use a smaller figsize and adjust subplot parameters
                fig, axes = plt.subplots(len(classes), n_samples, figsize=(8, 1.5 * len(classes)))
                fig.suptitle("Sample Images from Each Class", size=14)

                if len(classes) == 1:  # Handle the single-class case
                    axes = [axes]

                for i, class_name in enumerate(classes):
                    for j, img in enumerate(class_images[i]):
                        img = img.permute(1, 2, 0).numpy()
                        img = (img - img.min()) / (img.max() - img.min())
                        ax = axes[i, j] if len(classes) > 1 else axes[j] # Handle single class case.
                        ax.imshow(img)
                        ax.axis('off')
                        # Add class label as y-axis label, only for the first image in each row
                        if j == 0:
                            ax.text(-0.1, 0.5, class_name, transform=ax.transAxes,
                                    fontsize=10, va='center', ha='right', rotation='vertical') #Added rotation


                # Adjust layout for tighter spacing and title positioning
                plt.tight_layout(rect=[0, 0, 1, 0.95], h_pad=0.1, w_pad=0.1) # Added h_pad and w_pad
                plt.show()

            except Exception as e:
                print(f"Warning: Could not show sample images: {str(e)}")
        print("\nüñºÔ∏è Sample Images from Each Class")
        show_samples(train_loader, classes, n_samples)



        # Per-channel statistics
        try:
            print("\nüìä Per-channel Statistics")
            print("=" * 50)
            channels = ['Red', 'Green', 'Blue']
            for i, channel in enumerate(channels):
                channel_data = images[:, i, :, :].flatten()
                print(f"\nüî∏ {channel} Channel:")
                print(f"   Mean: {channel_data.mean():.3f}")
                print(f"   Std: {channel_data.std():.3f}")
                print(f"   Min: {channel_data.min():.3f}")
                print(f"   Max: {channel_data.max():.3f}")
        except Exception as e:
            print(f"Warning: Could not compute channel statistics: {str(e)}")

        # Data augmentation preview
        try:
            print("\nüîÑ Transform Pipeline:")
            print("=" * 30)
            for t in train_loader.dataset.transform.transforms:
                print(f"üî∏ {t.__class__.__name__}")
                if isinstance(t, torch.nn.Module):
                    print(f"   Parameters: {t.state_dict()}")
        except Exception as e:
            print(f"Warning: Could not inspect transform pipeline: {str(e)}")

    except Exception as e:
        print(f"Error in dataset exploration: {str(e)}")
        raise
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

def plot_distribution_with_stats(data, title, figsize=(15, 8)):
    """
    Enhanced helper function to create a distribution plot with a detailed statistics summary.
    Now uses subplots to properly display the statistical summary.
    """
    # Calculate key statistics
    mean = data.mean()
    std_dev = data.std()
    min_val = data.min()
    max_val = data.max()
    median = np.median(data)
    q1, q3 = np.percentile(data, [25, 75])
    skewness = stats.skew(data.flatten())
    kurtosis = stats.kurtosis(data.flatten())

    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize, gridspec_kw={'width_ratios': [3, 1]})

    # Define custom colors
    main_color = '#3498db'   # Blue
    mean_color = '#e74c3c'   # Red
    std_color = '#2ecc71'    # Green
    median_color = '#f39c12' # Orange

    # Plot histogram with KDE
    sns.histplot(data=data.flatten(), bins=50, kde=True, color=main_color,
                 edgecolor='white', alpha=0.7, stat='density', ax=ax1)

    # Add statistical markers
    ax1.axvline(mean, color=mean_color, linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')
    ax1.axvline(median, color=median_color, linestyle='--', linewidth=2, label=f'Median: {median:.2f}')
    ax1.axvline(mean + std_dev, color=std_color, linestyle='--', linewidth=2, label=f'Mean ¬± Std Dev: {std_dev:.2f}')
    ax1.axvline(mean - std_dev, color=std_color, linestyle='--', linewidth=2)

    # Highlight the area within one standard deviation
    kde = stats.gaussian_kde(data.flatten())
    x_range = np.linspace(min_val, max_val, 200)
    ax1.fill_between(x_range, kde(x_range),
                     where=(x_range >= mean-std_dev) & (x_range <= mean+std_dev),
                     color=std_color, alpha=0.1)

    # Customize histogram appearance
    ax1.set_title(title, fontsize=16, fontweight='bold', pad=15)
    ax1.set_xlabel('Values', fontsize=12, labelpad=10)
    ax1.set_ylabel('Density', fontsize=12, labelpad=10)
    ax1.grid(True, alpha=0.3, linestyle='--')

    # Improve legend
    ax1.legend(loc='upper right', frameon=True, fancybox=True, shadow=True)

    # Remove top and right spines
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)

    # ** Subplot for statistical summary **
    stats_text = (
      f"Statistical Summary:\n"
      f"‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ\n"
      f"- Mean: {mean:.2f}\n"
      f"- Median: {median:.2f}\n"
      f"- Std Dev: {std_dev:.2f}\n"
      f"- Min: {min_val:.2f}\n"
      f"- Max: {max_val:.2f}\n"
      f"- Q1: {q1:.2f}\n"
      f"- Q3: {q3:.2f}\n"
      f"- Skewness: {skewness:.2f}\n"
      f"- Kurtosis: {kurtosis:.2f}"
    )


    # Add text inside the second subplot with a clean box
    ax2.text(0.05, 0.5, stats_text, fontsize=12, verticalalignment='center', fontfamily='monospace',
             bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'))

    # Hide axes in the second panel
    ax2.axis('off')

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()

    return mean, std_dev, min_val, max_val, median, q1, q3, skewness, kurtosis
```

```python
def plot_cifar10_grid(test_loader, figsize=(15, 15), samples_per_class=8):
    """
    Creates a grid visualization of CIFAR-10 dataset images.

    Args:
        test_loader: PyTorch DataLoader containing CIFAR-10 test dataset
        figsize (tuple): Size of the output figure (width, height)
        samples_per_class (int): Number of samples to display for each class

    The grid will show samples_per_class rows, with each row containing
    one sample from each of the 10 CIFAR-10 classes.
    """
    import matplotlib.pyplot as plt
    import numpy as np

    # Get class names from the dataset
    classes = test_loader.dataset.classes
    num_classes = len(classes)

    # Create the figure with specified size
    plt.figure(figsize=figsize)

    # Add a main title to the figure
    plt.suptitle('CIFAR-10 Image Samples\nDisplaying {} samples per class'.format(samples_per_class),
                 fontsize=16, y=0.95)

    # Iterate through each class
    for class_idx, class_name in enumerate(classes):
        # Find all indices where the target matches the current class
        class_indices = np.flatnonzero(
            class_idx == np.array(test_loader.dataset.targets)
        )

        # Randomly sample indices for this class
        sample_indices = np.random.choice(
            class_indices,
            samples_per_class,
            replace=False
        )

        # Plot each sample for the current class
        for sample_num, idx in enumerate(sample_indices):
            # Calculate position in the grid
            plt_idx = sample_num * num_classes + class_idx + 1

            # Create subplot at the calculated position
            plt.subplot(samples_per_class, num_classes, plt_idx)

            # Display the image
            plt.imshow(test_loader.dataset.data[idx])
            plt.axis('off')

            # Add class name as title for the first row only
            if sample_num == 0:
                plt.title(class_name, fontsize=10)

    # Adjust the layout to prevent overlap and leave room for the main title
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

### Highlights for Model Evaluation

This function evaluates the performance of a model on a specific dataset by calculating accuracy as the primary metric. It provides a straightforward and quantitative measure of how well the model is predicting on that dataset.

* **`model.eval()`**

Configures the model to evaluation mode. This deactivates
training-specific behaviors or layers, such as:

**Dropout**: Deactivated, allowing all neurons to contribute fully during inference.

**Batch Normalization**: Uses the learned running estimates (mean and variance) without updating them, ensuring consistency in predictions during inference.

* **`torch.no_grad()`**

Disables gradient computation within this block of code. The benefits include:

**Memory Savings**: By not calculating or storing gradients, memory usage is reduced.

**Increased Speed**: Accelerates the evaluation process by skipping unnecessary backpropagation operations.

**Avoids Unnecessary Data Storage**: Prevents storing information that's only relevant during training, optimizing resources during validation or testing.

* **`scores = model(xi)`**

Feeds the inputs through the model to obtain preliminary predictions. These scores are the raw outputs of the model before applying activation functions like Softmax.



```python
def accuracy(model, loader):
    num_correct = 0
    num_total = 0
    model.eval()
    model = model.to(device=device)
    with torch.no_grad():
        for xi, yi in loader:
            xi = xi.to(device=device, dtype = torch.float32)
            yi = yi.to(device=device, dtype = torch.long)
            scores = model(xi)
            _, pred = scores.max(dim=1)
            num_correct += (pred == yi).sum()
            num_total += pred.size(0)
        return float(num_correct)/num_total
```

### Training Loop

### Highlights for Trainig Loop

* **`history = {'loss': [], 'val_acc': []}`**

Keep a record of the loss history and validation accuracy throughout the training epochs.

**`history['loss']`:** A list that will store the value of the loss function at the end of each epoch.

**`history['val_acc']`:** A list that will store the accuracy on the validation set at the end of each epoch.

* **`model = model.to(device=device)`**

Moves the model to the specified device (CPU or GPU). This ensures that both the model and the data are located on the same device, which is essential‚Äîespecially when using a GPU to accelerate training.

* **`model.train()`**

Activates training mode, which is essential for certain layers to function correctly. For example:

**Dropout Layers:** Randomly deactivate certain neurons during training to prevent overfitting.

**Batch Normalization Layers:** Update running statistics (mean and variance) during training but use them as fixed parameters during evaluation.


* **`cost = F.cross_entropy(input=scores, target=yi)`**

Calculates the loss using cross-entropy, which is suitable for classification tasks.

Measures the difference between the predicted class probabilities and the actual class labels.

Penalizes the model when the predicted probability for the actual class is low, encouraging the model to produce high probabilities for the correct classes


* **`optimiser.zero_grad()`**

Resets the gradients of all model parameters to avoid accumulation from the previous batch, which might lead to incorrect weight updates.

If not reset, accumulated gradients can cause the optimization algorithm to make larger updates than intended, potentially overshooting minima or causing instability.

* **`cost.backward()`**

Computes the gradients of the loss with respect to each model parameter.

Propagates the error backward through the network, calculating each parameter's contribution to the overall loss.

These gradients are stored in the .grad attribute of each parameter and are used by the optimizer to update the weights.

* **`optimiser.step()`**

Updates the model's weights using the calculated gradients.

The optimizer adjusts each parameter in the direction that minimizes the loss, based on its specific algorithm (e.g., SGD, Adam)

Gradients indicate both the direction and magnitude of parameter adjustments needed to minimize loss.


* **Importance of Tracking Loss and Accuracy**

By storing the loss and accuracy values at each epoch, you can plot learning curves to visualize and understand the training process. This helps identify issues like overfitting (where training loss decreases while validation loss increases) or underfitting (where both losses remain high).

Visualizing these metrics allows for better insights into how the model learns over time.

* **Hyperparameter Optimization:**

Adjusting parameters such as the **learning rate**, **batch size**, and **number of epochs** can significantly improve the model's performance.

**Learning Rate:** Controls how much the model weights are updated during training.

**Batch Size:** Determines the number of samples that propagate through the network before updating the weights.

**Number of Epochs:** Sets how many times the entire training dataset is passed forward and backward through the neural network.

### Sequential CNN

### Highlights for Secuential CNN

* **`channel1` and `channel2`**: These parameters determine the number of feature maps that the convolutional layers will produce. By increasing the number of channels, you allow the network to learn more complex and abstract features from the data.

* **`lr`**: This is the learning rate, which controls how much the network's weights are adjusted during each step of the training process. A smaller learning rate leads to more precise but slower convergence, while a larger one speeds up learning but may overshoot minima.

* **`epochs`**: Specifies the number of epochs, or complete passes over the training dataset. Each epoch means the model has seen all samples in the training set once.

### 1Ô∏è‚É£ First Convolutional Layer (Conv2d)

* **`in_channels`**: Indicates that the input has 3 channels, as it is an RGB image.

* **`out_channels`**: Specifies that this layer will generate 16 feature maps.

* **`kernel_size`**: The size of the filter used during convolution is 3x3 pixels.

* **`padding`**: A padding of 1 pixel is applied around the image. This technique adds extra pixels (usually zeros) at the edges before the convolution operation, thus maintaining the spatial dimensions of the image after convolution.

* **`ReLU`**: Applies the ReLU activation function, introducing non-linearity and helping the network learn more complex relationships.

### 2Ô∏è‚É£ Second Convolutional Layer (Conv2d)

* **`in_channels`**: Takes as input the 16 feature maps generated by the previous layer.

* **`out_channels`**: Generates 32 feature maps.

* Maintains **`kernel_size`** and **`padding`**: Uses the same filter size and padding as the previous layer.

* **`ReLU`**: The ReLU activation function is applied again to introduce non-linearity.

### 3Ô∏è‚É£ MaxPooling Layer (MaxPool2d)

* **`kernel_size`** and **`stride`**: Reduces the spatial dimensions of the feature maps by half. If the input is 32x32 pixels, after this layer it will be 16x16 pixels. This helps reduce dimensionality and focuses on the most prominent features.

### 4Ô∏è‚É£ Flattening (Flatten)

* Converts the 2D feature maps into a 1D vector to connect it to a fully connected layer.

### 5Ô∏è‚É£ Linear Layer (Linear):

* **`in_features`**: Receives as input the result of the feature maps after pooling, calculated as 16 (height) √ó 16 (width) √ó 32 (channels).

* **`out_features`**: Generates an output of 10 neurons, one for each of the 10 classes.

```python
# To do: Use PyTorch to implement a CNN model, accuracy should be better than
#the linear model
channel1 = 16
channel2 = 32
epochs = 50
lr = 5e-5
CNNmodel_1 = nn.Sequential(nn.Conv2d(in_channels=3, out_channels=channel1,
                                   kernel_size=3, padding=1),
                          nn.ReLU(),
                          nn.Conv2d(in_channels=channel1, out_channels=channel2,
                                    kernel_size= 3, padding=1),
                          nn.ReLU(),
                          nn.MaxPool2d(2, 2),
                          nn.Flatten(),
                          nn.Linear(in_features=16*16*channel2, out_features=10)
                                   )
optimiser = torch.optim.Adam(CNNmodel_1.parameters(), lr)
```

### Training Sequential CNN

```python
print("\n" + "="*50)
print("Sequential CNN Training Initiated".center(50))
print("="*50 + "\n")

historyCNN_1 = train(CNNmodel_1, optimiser, epochs)

print("\n" + "="*50)
print("Training Complete!".center(50))
print(f"Final Accuracy: {historyCNN_1['val_acc'][-1]*100:.2f}%".center(50))
print("="*50 + "\n")


```

```python
plot_training(historyCNN_1)
```

### Improving the CNN Model

* **Adding 2 Convolutional Layers**: Introducing two additional convolutional layers increases the network's depth, allowing it to learn more complex and deeper features from the data.

* **Adding 2 Channels (64 and 128)**: By incorporating channels with 64 and 128 filters, the model can capture more intricate and detailed patterns, enhancing its learning capacity.

* **Including Batch Normalization After Each Convolutional Layer**: Adding a Batch Normalization layer after each convolutional layer helps reduce internal covariate shift, which improves training efficiency for deeper models.

**Original architecture**

$$2\,Convolutional\,Layers ‚Üí ReLU ‚Üí MaxPooling ‚Üí Flatten ‚Üí Linear\,Layer$$

**Improved Architecture**

$$4\,Convolutional\,Layers (with\,BatchNorm\,and\,ReLU) ‚Üí MaxPooling ‚Üí Flatten ‚Üí Linear\,Layer$$

**Benefits of the Improvements**

* **`Greater Model Depth and Width`**: Allows the network to capture more complex and abstract features.

* **`Incorporation of Batch Normalization`**: Enhances training stability and speed by normalizing the inputs of each layer.

* **`Increased Learning Rate`**: Accelerates convergence thanks to batch normalization, which allows for a higher learning rate without compromising stability.


```python
input_channels = 3
channel1 = 16
channel2 = 32
channel3 = 64
channel4 = 128
output_features = 10

epochs = 50
lr = 1e-3

historyCNN_2 = nn.Sequential(nn.Conv2d(in_channels=input_channels, out_channels=channel1,kernel_size=3, padding=1),
                          nn.BatchNorm2d(channel1),
                          nn.ReLU(),

                          nn.Conv2d(in_channels=channel1, out_channels=channel2,kernel_size=3, padding=1),
                          nn.BatchNorm2d(channel2),
                          nn.ReLU(),

                          nn.Conv2d(in_channels=channel2, out_channels=channel3,kernel_size=3, padding=1),
                          nn.BatchNorm2d(channel3),
                          nn.ReLU(),

                          nn.Conv2d(in_channels=channel3, out_channels=channel4,kernel_size=3, padding=1),
                          nn.BatchNorm2d(channel4),
                          nn.ReLU(),

                          nn.MaxPool2d(2, 2),
                          nn.Flatten(),

                          nn.Linear(in_features=16*16*channel4, out_features=output_features)
)

optimiser = torch.optim.Adam(historyCNN_2.parameters(), lr)
```

```python
print("\n" + "="*50)
print("Sequential CNN Training Initiated".center(50))
print("="*50 + "\n")

historyCNN2 = train(historyCNN_2, optimiser, epochs)

print("\n" + "="*50)
print("Training Complete!".center(50))
print(f"Final Accuracy: {historyCNN2['val_acc'][-1]*100:.2f}%".center(50))
print("="*50 + "\n")
```

```python
plot_training(historyCNN2)
```
