```python
# Del primer archivo obtenemos una matriz con 3 evaluaciones de los restaurantes:
# general, comida y servicio.
# Las evaluaciones pueden ser 0, 1 o 2, siendo 0 la menor calificación y 2 la
# mayor calificación:

data1.head()
```

```python
# De data1 no requeriremos "rating" y de data2 solo necesitamos "placeID" y "name":
# Definimos la lista y matrices con los factores que necesitamos:

lista_data1 = ['userID','placeID','food_rating','service_rating']
lista_data2 = ['placeID','name']

data1a = data1[lista_data1]
data2a = data2[lista_data2]
```

## **Ejercicio - 2**

```python
# Define el DataFrame que conjunte la información de las dos DataFrame data1a
# y data2a en uno solo, a través de la columna común "placeID" y que
# llamaremos "df_combinado":



# ************* Inlcuye aquí tu código:*****************************

# Combina los DataFrames 'data1a' y 'data2a' utilizando 'placeID' como clave común para crear 'df_combinado'
df_combinado = pd.merge(data1a, data2a, on='placeID')

# *********** Aquí termina la sección de agregar código *************


# Despleguemos la dimensión y los primeros renglones de este DataFrame:

print(df_combinado.shape)
df_combinado.head()
```

## **Ejercicio - 3**

```python
#    Define la matriz de utilidad cuyos renglones sean los nombres de los
#    restaurantes, las columnas los IDs de los usuarios y las entradas la
#    evaluación de la comida (food_rating). La llamaremos "UtMx_food".


# ************* Inlcuye aquí tu código:*****************************


# Creamos la matriz de utilidad 'UtMx_food' donde:
# - Los renglones son los nombres de los restaurantes ('name')
# - Las columnas son los IDs de los usuarios ('userID')
# - Las entradas son la evaluación de la comida ('food_rating')
# Utilizamos el método pivot_table para reorganizar los datos de 'df_combinado'

UtMx_food = df_combinado.pivot_table(values='food_rating', index='name', columns='userID', fill_value=0)



# *********** Aquí termina la sección de agregar código *************


print('Dimensión de la matriz de Utilidad sobre la evaluación de la comida:')
print('(restaurantes, usuarios) =', (UtMx_food.shape))
UtMx_food.head()
```

## **Ejercicio - 4**

### **En la Factorización SVD la cantidad de valores singulares será menor o igual al menor valor de los renglones o columnas de la matriz de utilidad UtMx.**

### **La factorización SVD de una matriz $A$ tiene la forma:**

$A_{m\times n} = U_{m\times m}\Sigma_{m\times n}V_{n\times n}^T$

```python
# a. Aplica la transformación SVD para obtener la matriz de las variables
#    latentes de los restaurantes, en relación a la manera en que los usuarios
#    evaluaron la comida. Deberás obtener la factorización SVD con el máximo número
#    de componentes y por lo tanto el máximo de valores singulares posibles de acuerdo
#    a este criterio. Así, de acuerdo al resultado obtenido en el Ejercicio-3, asigna
#    a la variable "nc_food" el número entero adecuado para llegar a obtener el
#    resultado pedido.


# ************* Inlcuye aquí tu código:**********************************************


# Obtenemos las dimensiones de la matriz de utilidad UtMx_food
num_restaurantes, num_usuarios = UtMx_food.shape

# El número máximo de valores singulares es el menor entre el número de renglones y columnas
nc_food = min(num_restaurantes, num_usuarios)

# *********** Aquí termina la sección de agregar código *****************************

print('Total de valores singulares basados en evaluación de la comida:', nc_food)
```

```python
# b. Lleva a cabo la factorización SVD truncada en la matriz de utilidad
#    obtenida previamente y con respecto al número de componentes dada
#    por la variable nc_food.


# ************* Inlcuye aquí tu código:********************************


# Aplicamos la factorización SVD truncada a la matriz de utilidad 'UtMx_food'
# utilizando el número de componentes 'nc_food' calculado previamente

SVD_food = TruncatedSVD(n_components=nc_food)
SVD_food.fit(UtMx_food)


# *********** Aquí termina la sección de agregar código ***************


# Con la factorización SVD truncada obtenida, determinemos la cantidad
# de componentes que explican un 90% de la variabilidad acumulada de
# cada componente:

for j in range(nc_food):
  if SVD_food.explained_variance_ratio_[0:j].sum() > 0.90:
    break

# Usaremos esta cantidad de componentes para las recomendaciones basadas
# en la calificación de la comida:
N_food = j-1


print('Valor de truncamiento al 90% de dicha variabilidad:', N_food)
```

```python
# Usando la métrica de correlación de Pearson, obtenemos ahora las 10 mejores
# recomendaciones de este modelo no supervisado con base a la información de
# alguien que desea obtener las similitudes con el "Restaurante Pueblo Bonito"
# y con la cota de truncamiento obtenido en el ejercicio anterior.
# Para ello obtenemos las 10 mejores correlaciones positivas a continuación.

# Factorización SVD:
SVD_food = TruncatedSVD(n_components = N_food)
resultant_matrix_food = SVD_food.fit_transform(UtMx_food)

# Matriz de correlación de Pearson:
corr_mat_food = np.corrcoef(resultant_matrix_food)

# Restaurante de referencia:
restaurante_de_referencia = "Restaurante Pueblo Bonito"
nombres_rest = UtMx_food.T.columns  # nombres de restaurantes
idx_rest = list(nombres_rest).index(restaurante_de_referencia)
corr_rest = corr_mat_food[idx_rest] # Vector de Correlación del RinconHuasteco

# Buscando las correlaciones positivas:
idx = (corr_rest>0)
mejores_sim_food = list()
for i in range(len(nombres_rest[idx])):
  mejores_sim_food.append((corr_rest[idx][i], nombres_rest[idx][i]))

print('Total de similaridades positivas encontradas:', len(mejores_sim_food))

print('Algunos de los resultados encontrados:')
mejores_sim_food[0:5]
```

## **Ejercicio - 5**

```python
# Ordenar la lista de recomendaciones "mejores_sim_food" encontrada en el paso
# anterior de manera descendente. Llamaremos "mejores_sim_food_ordenadas" a
# dicha lista.

# ************* Inlcuye aquí tu código:*****************************



# Ordenar la lista de tuplas por el primer elemento (rating) de mayor a menor
mejores_sim_food_ordenadas = sorted(mejores_sim_food, key=lambda x: x[0], reverse=True)



# *********** Aquí termina la sección de agregar código *************


# Desplegamos las 10 mejores similitudes encontradas de manera descendente:
print('Similitudes con base a la evaluación de la comida con mayores valores de correlación:')
for k in range(1,11):
  print('%d> %s' % (k, mejores_sim_food_ordenadas[k]))
```

## **Ejercicio - 6**

### **De manera análoga y usando ahora la evaluación de un restaurante con respecto a su servicio (service_rating), encuentra ahora los diez restaurantes con mayor similaridad al mismo restaurante anterior llamado "Restaurante Pueblo Bonito" y con respecto también a la misma métrica de similaridad de Pearson.**

### **Deberás usar la cantidad de componentes necesarias para que la varabilidad de la factorización SVD quede explicada en un 90%.**

### **a.  Para ello define primero la matriz de utilidad, UtMx_service, donde los renglones son los nombres de los restaurantes, las columnas los usuarios y las entradas las evaluaciones con respecto al servicio recibido en el restaurante.**





```python
# a. Define la matriz de utilidad cuyos renglones sean los nombres de los
#    restaurantes, las columnas los IDs de los usuarios y las entradas la
#    evaluación del servicio (service_rating). La llamaremos "UtMx_service".


# ************* Inlcuye aquí tu código:*****************************


# Creamos la matriz de utilidad 'UtMx_service' donde:
# - Los renglones son los nombres de los restaurantes ('name')
# - Las columnas son los IDs de los usuarios ('userID')
# - Las entradas son la evaluación del servicio ('service_rating')
# Utilizamos el método pivot_table para reorganizar los datos de 'df_combinado'

UtMx_service = df_combinado.pivot_table(values='service_rating', index='name', columns='userID', fill_value=0)



# *********** Aquí termina la sección de agregar código *************




print('Dimensión de la matriz de Utilidad sobre la evaluación del servicio:')
print('(restaurantes, usuarios) =', (UtMx_service.shape))
UtMx_service.head()
```

```python
# b. Incluye el código necesario para obtener y desplegar los 10 restaurantes con
#    mayor similaridad de Pearson al "Restaurante Pueblo Bonito", y usando solo
#    el 90% de las componentes que explican la varianza de la matriz UtMx_service.
#    Desplegarlos en orden descendente, sin incliur el restaurante de referencia.

# ************* Inlcuye aquí tu código:*****************************

# Paso 1: Determinar el número máximo de componentes para SVD
num_restaurantes, num_usuarios = UtMx_service.shape
nc_service = min(num_restaurantes, num_usuarios)

# Paso 2: Aplicar SVD truncado con el número máximo de componentes
SVD_service = TruncatedSVD(n_components=nc_service)
SVD_service.fit(UtMx_service)

# Paso 3: Calcular la varianza acumulada para determinar el número de componentes necesarios
varianza_acumulada = np.cumsum(SVD_service.explained_variance_ratio_)
N_service = np.argmax(varianza_acumulada >= 0.90) + 1  # Sumamos 1 porque los índices comienzan en 0

print('Número de componentes para explicar el 90% de la varianza:', N_service)

# Paso 4: Reentrenar el SVD truncado con el número reducido de componentes
SVD_service = TruncatedSVD(n_components=N_service)
matriz_latente_service = SVD_service.fit_transform(UtMx_service)

# Paso 5: Obtener el índice del "Restaurante Pueblo Bonito"
nombre_restaurante = "Restaurante Pueblo Bonito"

# Verificamos que el restaurante exista en el índice
if nombre_restaurante in UtMx_service.index:
    indice_restaurante = UtMx_service.index.get_loc(nombre_restaurante)
else:
    raise ValueError(f'El restaurante "{nombre_restaurante}" no se encuentra en la matriz de utilidad.')

# Paso 6: Obtener el vector latente del restaurante de referencia
vector_restaurante = matriz_latente_service[indice_restaurante]

# Paso 7: Calcular la similaridad de Pearson entre el restaurante de referencia y los demás
mejores_sim_service = []

for idx, nombre in enumerate(UtMx_service.index):
    if nombre != nombre_restaurante:
        vector_otro_restaurante = matriz_latente_service[idx]
        # Calculamos el coeficiente de correlación de Pearson
        similitud, _ = pearsonr(vector_restaurante, vector_otro_restaurante)
        # Agregamos la similitud y el nombre del restaurante a la lista
        mejores_sim_service.append((similitud, nombre))

# Paso 8: Ordenar la lista de similitudes de mayor a menor
mejores_sim_service_ordenadas = sorted(mejores_sim_service, key=lambda x: x[0], reverse=True)

# Paso 9: Desplegar las 10 mejores similitudes encontradas de manera descendente
print(f'Restaurantes más similares a "{nombre_restaurante}" basados en la evaluación del servicio:')
for k in range(10):
    similitud, nombre_restaurante_similar = mejores_sim_service_ordenadas[k]
    print(f"{k+1}> {nombre_restaurante_similar}: similitud = {similitud:.4f}")


# *********** Aquí termina la sección de agregar código *************
```

```python
# b. Incluye el código necesario para obtener y desplegar los 10 restaurantes con
#    mayor similaridad de Pearson al "Restaurante Pueblo Bonito", y usando solo
#    el 90% de las componentes que explican la varianza de la matriz UtMx_service.
#    Desplegarlos en orden descendente, sin incluir el restaurante de referencia.

# ************* Incluye aquí tu código: *****************************

# Determinar el número máximo de componentes para SVD
num_restaurantes, num_usuarios = UtMx_service.shape
nc_service = min(num_restaurantes, num_usuarios)

# Aplicar SVD truncado con el número máximo de componentes
SVD_service = TruncatedSVD(n_components=nc_service)
SVD_service.fit(UtMx_service)

# Calcular la varianza acumulada para determinar el número de componentes necesarios
varianza_acumulada = np.cumsum(SVD_service.explained_variance_ratio_)
N_service = np.argmax(varianza_acumulada >= 0.90) + 1  # Sumamos 1 porque los índices comienzan en 0

print(f'Número de componentes para explicar el 90% de la varianza: {N_service}')

# Reentrenar el SVD truncado con el número reducido de componentes
SVD_service = TruncatedSVD(n_components=N_service)
matriz_latente_service = SVD_service.fit_transform(UtMx_service)

# Obtener el índice del "Restaurante Pueblo Bonito"
nombre_restaurante = "Restaurante Pueblo Bonito"

# Verificamos que el restaurante exista en el índice
if nombre_restaurante in UtMx_service.index:
    indice_restaurante = UtMx_service.index.get_loc(nombre_restaurante)
else:
    raise ValueError(f'El restaurante "{nombre_restaurante}" no se encuentra en la matriz de utilidad.')

# Obtener el vector latente del restaurante de referencia
vector_restaurante = matriz_latente_service[indice_restaurante]

# Calcular la similaridad de Pearson entre el restaurante de referencia y los demás
mejores_sim_service = []

for idx, nombre in enumerate(UtMx_service.index):
    if nombre != nombre_restaurante:
        vector_otro_restaurante = matriz_latente_service[idx]
        # Calculamos el coeficiente de correlación de Pearson
        similitud, _ = pearsonr(vector_restaurante, vector_otro_restaurante)
        # Agregamos el nombre del restaurante y la similitud a la lista
        mejores_sim_service.append({'Restaurante': nombre, 'Similitud': similitud})

# Convertir la lista a un DataFrame y ordenarlo
df_similitud = pd.DataFrame(mejores_sim_service)
df_similitud = df_similitud.sort_values(by='Similitud', ascending=False)

# Definimos una función para aplicar estilos condicionales con colores
def resaltar_similitud(val):
    if val > 0.8:
        color = 'green'  # Mantiene el verde para altas similitudes
    elif val > 0.5:
        color = 'orange'  # Mantiene el naranja para similitudes medias
    else:
        color = 'lightcoral'  # Rojo suave para bajas similitudes
    return f'color: {color}'

# Seleccionamos las primeras 10 filas
top_10 = df_similitud.head(10).reset_index(drop=True)

# Aplicamos el estilo
styled_top_10 = top_10.style.map(resaltar_similitud, subset=['Similitud']) \
                            .set_properties(**{'text-align': 'left'}) \
                            .set_table_styles([{'selector': 'th', 'props': [('text-align', 'left')]}])

# Mostramos el DataFrame con estilos y colores suaves
display(HTML(f"<h3 style='color: skyblue;'>Restaurantes más similares a '{nombre_restaurante}' basados en la evaluación del servicio:</h3>"))
display(styled_top_10)

# *********** Aquí termina la sección de agregar código *************
```

Esta semana, nos hemos sumergido en técnicas fundamentales de **reducción de dimensionalidad** y su aplicación en **sistemas de recomendación**, áreas esenciales en el campo del **Machine Learning**. La comprensión y manejo de estas técnicas son cruciales por varias razones:

- **Manejo Eficiente de Datos**: En la era del Big Data, los conjuntos de datos suelen ser de alta dimensionalidad. Técnicas como la **Descomposición en Valores Singulares (SVD)** y el **Análisis de Componentes Principales (PCA)** permiten reducir la complejidad de los datos sin perder información significativa, facilitando el almacenamiento y procesamiento eficientes.

- **Descubrimiento de Patrones Latentes**: La reducción de dimensionalidad ayuda a identificar variables latentes que no son observables directamente pero que influyen en los datos. Esto es especialmente útil en sistemas de recomendación, donde podemos descubrir preferencias ocultas de los usuarios y tendencias en los productos o servicios.

- **Mejora del Rendimiento de Modelos**: Al eliminar el ruido y las redundancias en los datos, los modelos de Machine Learning pueden entrenarse más rápido y con mayor precisión, evitando el sobreajuste y mejorando la generalización.

- **Aplicaciones Prácticas y Cotidianas**: Los sistemas de recomendación están presentes en múltiples plataformas que utilizamos diariamente, como servicios de streaming, tiendas en línea y redes sociales. Entender cómo funcionan nos permite no solo mejorar estas herramientas sino también innovar en nuevas aplicaciones que personalicen y optimicen la experiencia del usuario.

- **Fundamento para Técnicas Avanzadas**: La reducción de dimensionalidad es un pilar para otras técnicas más avanzadas en Machine Learning, como el procesamiento de lenguaje natural, el análisis de imágenes y el aprendizaje profundo. Dominar estos conceptos sienta las bases para abordar problemas más complejos y desafiantes.

Esta semana ha reforzado la importancia de combinar teoría y práctica, y cómo herramientas matemáticas robustas pueden tener un impacto significativo en soluciones reales que afectan a millones de usuarios en todo el mundo.
