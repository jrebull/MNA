## **Ejercicio 5**

```python
# Transformaciones que se aplicar√°n a las variables num√©ricas usando la clase Pipeline de sklearn:

# ************* Inlcuye aqu√≠ tu c√≥digo:*****************************

# Variables num√©ricas:
numericas_pipe = Pipeline([("impute", SimpleImputer(strategy="median")),
                           ("standardize", MinMaxScaler())])
numericas_pipe_nombres = ['duration', 'amount','age', 'people_liable']

# Variables categ√≥ricas-Nominales:
nominales_pipe = Pipeline([("impute", SimpleImputer(strategy="most_frequent")),
                           ("encoder", OneHotEncoder(handle_unknown="ignore", drop = "first"))])
nominales_pipe_nombres = ['status','credit_history','purpose','savings','personal_status_sex',
                          'other_debtors','other_installment_plans','housing','telephone','foreign_worker']#quite credict_risk

# Variables categ√≥ricas-ordinales:
ordinales_pipe = Pipeline([("impute", SimpleImputer(strategy="most_frequent")),
                           ("encoder", OrdinalEncoder())])
ordinales_pipe_nombres = ['employment_duration','installment_rate','present_residence','property','number_credits','job']


# Conjuntas las transformaciones de todo tipo de variable y
# deja sin procesar aquellas que hayas decidido no transformar:

columnasTransformer = ColumnTransformer(transformers=[('num', numericas_pipe, numericas_pipe_nombres),
                                                      ('ord', ordinales_pipe, ordinales_pipe_nombres),
                                                      ('nom', nominales_pipe, nominales_pipe_nombres)], remainder='passthrough')

# *********** Aqu√≠ termina la secci√≥n de agregar c√≥digo *************
```

## **Ejercicio 6**

```python
# Como se va a utilizar Validaci√≥n-Cruzada, concatena los conjuntos de entrenamiento y prueba
# en uno nuevo conjunto aumentado que llamaremos trainval para utilizar como entrenamiento:

# ************* Inlcuye aqu√≠ tu c√≥digo:**************************

Xtraintest = pd.concat([Xtrain, Xtest], axis=0)
ytraintest = pd.concat([ytrain, ytest], axis=0)

# *********** Aqu√≠ termina la secci√≥n de agregar c√≥digo *************

# Veamos cu√°ntas variables nuevas se introducen con las transformaciones One-Hot-Encoding:
Xtmp = Xtraintest.copy()
tmp = columnasTransformer.fit_transform(Xtmp)
print("Dimensi√≥n de las variables de entrada ANTES de las transformaciones:", Xtmp.shape)
print("Dimensi√≥n de las variables de entrada DESPU√âS de las transformaciones:", tmp.shape)
```

## **Ejercicio 7**

```python
# Definimos a continuaci√≥n la funci√≥n que llamamos "mis_modelos" que incluye
# todos los modelos que deseamos comparar en el ejercicio.


def mis_modelos():
  modelos, nombres = list(), list()


  # ************* Inlcuye aqu√≠ tu c√≥digo:**************************
  #
  # Deber√°s incluir en cada modelo los argumentos que consideres
  # adecuados para que cada uno converja y no est√© sobreentrenado
  # con respecto a la m√©trica de la exatitud (accuracy).


  # Regresi√≥n Log√≠stica - Logistic Regression-LR:
  # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

  modelos.append(LogisticRegression(penalty='elasticnet',
                                    solver = 'saga',
                                    tol=0.0001,     # tolerancia
                                    fit_intercept = True,
                                    max_iter = 2000,
                                    C = 10,
                                    l1_ratio= 0.34,
                                    random_state=1))
  nombres.append('LR')




  # k-Vecinos m√°s Cercanos : k-Nearest-Neighbors-kNN:
  # https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html

  modelos.append(KNeighborsClassifier(n_neighbors=20,
                                      p = 1,
                                      leaf_size = 35,
                                      n_jobs = -1,
                                      weights = "uniform") )
  nombres.append('kNN')




  # √Årbol de decisiones-DecisionTree-DT:
  # https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

  modelos.append(DecisionTreeClassifier(criterion = "entropy",
                                        max_depth = 10,
                                        min_samples_split = 15,
                                        random_state = 1 ))
  nombres.append('DTree')




  # Bosque Aleatorio-RandomForest-RF:
  # https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

  modelos.append(RandomForestClassifier(n_estimators = 200,
                                        criterion = "entropy",
                                        max_features = "sqrt",
                                        random_state =  1,
                                        max_depth=10))
  nombres.append('RF')




  # XGBoosting:
  # https://xgboost.readthedocs.io/en/latest/python/python_api.html#xgboost.XGBClassifier
  # https://xgboost.readthedocs.io/en/stable/parameter.html

  modelos.append(XGBClassifier(random_state  = 1 ,
                               tree_method="hist",
                               learning_rate=0.05,
                               max_depth=7))
  nombres.append('XGBoost')




  # Red neuronal de Perceptr√≥n Multicapa-MLP:
  # https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

  modelos.append(MLPClassifier(solver = "adam",
                               random_state  = 1 ,
                               hidden_layer_sizes = 32,
                               activation = "identity",
                               max_iter = 2500,
                               alpha = 0.001))
  nombres.append('MLP')




  # M√°quina de Vectores de Soporte-SVM:
  # https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

  modelos.append(SVC(C = 14,
                     kernel = "linear",
                     tol = 0.0001))
  nombres.append('SVM')

  return modelos, nombres


# T√©cnica de submuestreo (undersampling) y/o sobremuestreo (versampling) utilizada:
mi_uoSampling = KMeansSMOTE(random_state=1, k_neighbors = 6, n_jobs = -1) ##BorderlineSMOTE(random_state=1, k_neighbors = 10) #SMOTE(random_state=1, k_neighbors = 2)

Xtv_uo, ytv_uo = mi_uoSampling.fit_resample(Xtraintest, ytraintest)



# *********** Aqu√≠ termina la secci√≥n de agregar c√≥digo *******************





# Entrenamos cada uno de los modelos y desplegamos la m√©tricas de Train y Val.

# NOTA: Observa que el m√©todo de Validaci√≥n-Cruzada llama  a los resultados
#       de "validation" como "test":

modelos, nombres = mis_modelos()
resultados = list()

for i in range(len(modelos)):

  # Definimos nuestro pipeline con las transformaciones y los modelos:
  pipeline = Pipeline(steps=[('ct',columnasTransformer),('m',modelos[i])])

  # Aplicaremos validaci√≥n-cruzada:
  micv = RepeatedStratifiedKFold(n_splits=5,
                                 n_repeats=3,
                                 random_state=5     # agreguemos una semilla para estabilizar resultados.
                                 )


  # Definimos las m√©tricas que desamos recuperar:
  mismetricas = {'miaccuracy':'accuracy','miprecision':'precision','mirecall':'recall',
                 'mifi':'f1','miauc':'roc_auc','migmean':make_scorer(geometric_mean_score)}

  # Llevamos a cabo el entrenamiento:
  scores = cross_validate(pipeline,
                          Xtv_uo,
                          ytv_uo,
                          scoring=mismetricas,
                          cv=micv,
                          return_train_score=True,
                          )

  # Guardemos el resultado de cada modelo para an√°lisis posteriores.
  resultados.append(scores)

  # Desplegamos los valores de las m√©tricas para verificar si no hay
  # subentrenamiento o sobreentrenamiento:
  print('>> %s' % nombres[i])
  for j,k in enumerate(list(scores.keys())):
    if j>1:
      print('\t %s %.3f (%.3f)' % (k, np.mean(scores[k]),np.std(scores[k])))
```

### **Conclusiones Finales de la Actividad** :

En esta actividad se analizaron diferentes modelos y el uso de submuestreo o sobre-muestreo para casos donde las clases no estan balanceadas.

Entre las ventajas del sobre-muestreo esta que no se pierde informaci√≥n, en el caso de dataset peque√±os previene que se pierdan datos de la clase mayoritaria. Este tipo de t√©cnica agrega datos sint√©ticos a la clase minoritaria. Sin embargo, al aumentar el conjunto de datos los tiempos de entrenamiento y uso de memoria ser√°n mayores, adem√°s de un riesgo a sobreajuste del modelo.

En el caso del submuestreo, se reduce el tama√±o del dataset, lo que beneficia los tiempos de entrenaamiento y reduce a su vez el uso dde memoria. Para grandes vol√∫menes de datos se puede utilizar ya que no hay p√©rdida de datos significativa. Aun as√≠ esta el riesgo de perder informaci√≥n valiosa, por lo que no se sugiere para conjuntos de datos peque√±os.


Los modelos implementados fueron:


* **Logistic regression**: Este algoritmo de aprendizaje supervisado es utilizado para problemas de clasificcaci√≥n, predice las probabilidades de un objetivo binario seg√∫n las variables predictoras (X) realizando un ajuste lineal con una salida de 0 o 1.

* **K-Nearest Neighbors (KNN)**:Este algoritmo es utilizado para an√°lisis de clasificaci√≥n y regresi√≥n lineal. Predice la variable de salida basada en las K vecinas m√°s cercanas en el conjunto de entrenamiento. El valor de k determina el n√∫mero de vecinos a considerar a la hora de hacer la predicci√≥n.

* **Modelo de √°rbol de decisi√≥n**: Tambi√©n se trata de un algoritmo para an√°lisis de clasificacci√≥n y regresi√≥n. Divide de forma recurrente los datos en subconjuntos m√°s peque√±os para crear el arbol de decisi√≥n.

* **Random Forest**: Este algoritmo ajusta una serie de clasificadores de √°rboles de decisi√≥n en varias submuestras y utiliza promedios para mejorar la precisi√≥n y controlar el sobreajuste.

* **Extreme Gradient Boosting (XGBoost)**: O refuerzo de gradientes extremo, es un m√©todo basado en √°rboles de decisi√≥n. Realiza una predicci√≥n inicial y crea un arbol de decisi√≥n, la salida del √°rbol se utiliza para construir otro √°rbol  hasta que los residuales dejan de reducirse, o bien el n√∫mero de veces especificado.

* **Red neuronal de Perceptr√≥n Multicapa-MLP**: Las redes neuronales son modelos inspirados en el cerebro humano, consisten en multiples capas con varias neuronas cada una, capaces de aprender funciones complejas.

* **Support Vector Machines (SVM)**: Este algoritmo es usado principalmente para problemas de clasificaci√≥n pero tambien puede usarse en problemas de regresi√≥n. Clasifica los datos encontrando el hiperplano √≥ptimo que separa al m√°ximo las diferentes clases en los datos de entrada.
