## üìä Introduction to Word Embeddings

Word embeddings are dense vector representations of words in a continuous vector space where semantically similar words are positioned close to each other. Unlike sparse one-hot encodings, embeddings capture meaningful relationships between words and enable powerful NLP applications.

### üåê GloVe: Global Vectors for Word Representation

GloVe is an unsupervised learning algorithm that creates word vectors by analyzing global word-word co-occurrence statistics from a corpus. These pre-trained vectors encode semantic and syntactic information about words, making them valuable for various NLP tasks.

### üìà Visualizing Word Embeddings

Since word embeddings typically exist in high-dimensional spaces (50-300 dimensions), we use dimensionality reduction techniques to visualize them:

- **PCA (Principal Component Analysis)** üìâ: A linear technique that projects data onto axes of maximum variance
- **t-SNE** üîç: A non-linear technique that preserves local similarities, ideal for visualizing word clusters

### üß© Applications of Word Embeddings

**Word Similarity** üëØ‚Äç‚ôÄÔ∏è: Finding similar words by calculating vector distances or similarities (e.g., cosine similarity)

**Word Analogies** ‚öñÔ∏è: Solving relationships like "king - man + woman ‚âà queen" by leveraging the algebraic properties of word vectors

# üìñ Creating the Embeddings Dictionary

## üéØ Purpose
This function transforms the raw GloVe text file into a structured dictionary for efficient word vector lookups:
- **Keys**: Individual words (strings)
- **Values**: Corresponding embedding vectors (numpy arrays)

## üîÑ Process Flow
```
Raw Text File ‚Üí Parse Line by Line ‚Üí Extract Word & Vector ‚Üí Store in Dictionary
```

## üìù Detailed Steps

### 1Ô∏è‚É£ Initialize Empty Dictionary
```python
embeddings_dict = {}
```

### 2Ô∏è‚É£ Open & Process File
- Opens the text file with proper UTF-8 encoding
- Uses `tqdm` to display a progress bar during processing
- Processes millions of word vectors efficiently

### 3Ô∏è‚É£ Line-by-Line Processing
For each line in the file:
```
"cat 0.123 0.456 0.789 ... (47 more values)"
 ‚Üì     ‚Üì     ‚Üì     ‚Üì
 word  dim1  dim2  dim3 ...
```

- **Split Line**: `values = line.strip().split()`
- **Extract Word**: `word = values[0]` (first element)
- **Extract Vector**: `coefs = np.asarray(values[1:], dtype='float32')` (remaining elements)
- **Store in Dictionary**: `embeddings_dict[word] = coefs`

### 4Ô∏è‚É£ Error Handling & Validation

The function includes comprehensive error checking:

- ‚úÖ **Dimension Validation**: Confirms each vector has exactly 50 elements
- ‚úÖ **Value Validation**: Ensures all vector elements can be converted to floating-point numbers
- ‚úÖ **File Existence**: Verifies the GloVe file exists at the specified path
- ‚úÖ **Exception Catching**: Handles unexpected errors with informative messages

## üí° Why This Matters

Converting to a dictionary structure provides key advantages:
- **O(1) lookup time** for finding word vectors
- **Memory efficiency** through numpy arrays
- **Easy integration** with subsequent embedding operations
- **Direct access** for word similarity and analogy calculations

## üîç Example Usage
```python
# After dictionary creation

print(f"Dimensions: {vector.shape}")  # Should be (50,)
```

````python
def create_emb_dictionary(path, emb_dim=None):
    """
    Loads GloVe embeddings from a text file into a dictionary.

    Args:
        path (str): Path to the GloVe embedding file.
        emb_dim (int, optional): Expected embedding dimension.  If provided,
            performs a dimension check and prints warnings for mismatches.
            Defaults to None (no dimension check).

    Returns:
        dict: A dictionary where keys are words (str) and values are their
            corresponding embedding vectors (numpy arrays). Returns an empty
            dictionary on error (file not found, etc.).
    """
    dictionary = {}
    try:
        with open(path, mode='r', encoding="utf-8") as file:
            for line in tqdm(file, desc="Loading embeddings", unit=" lines"):
                values = line.strip().split()   # Use strip() for cleaner line processing (removes undesired characters)
                if not values:                  # Skip empty lines
                    continue
                word = values[0]
                try:
                    coefs = np.asarray(values[1:], dtype='float32')
                    if emb_dim is not None and len(coefs) != emb_dim:
                        print(f"\nWarning: Skipping line. Expected dimension {emb_dim}, got {len(coefs)} for '{word}'")
                        continue                # Skip lines with incorrect dimensions
                    dictionary[word] = coefs
                except ValueError:
                    print(f"\nWarning: Skipping line with invalid values for word '{word}'.")
                    continue                    # Skip lines with invalid numeric values
    except FileNotFoundError:
        print(f"\nError: File not found at {path}")
        return {}                               # Return empty dictionary on file not found
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
        return {}

    return dictionary
````

# üíæ Saving & Loading Embeddings with Pickle

## What is Pickle?

The `pickle` module is Python's built-in serialization tool that converts complex Python objects (like our embeddings dictionary) into byte streams and back again. Think of it as "freezing" your data structures so they can be stored and "thawed" later exactly as they were.

## Why Use Pickle with Word Embeddings?

Our GloVe embeddings dictionary contains **400,000+ words**, each with a **50-dimensional vector**. Processing this from the original text file:
- ‚è±Ô∏è Takes significant time (often minutes)
- üßÆ Requires repeated parsing and memory allocation
- üîÑ Needs to be done every time you restart your notebook

## How Pickle Helps:

```python
# Saving our processed dictionary
pickle.dump(embeddings_dict, open('glove_embeddings.pkl', 'wb'))

# Loading the dictionary later
embeddings_dict = pickle.load(open('glove_embeddings.pkl', 'rb'))
```

## üåü Key Benefits:

- **‚ö° Lightning-fast loading**: Restore your embeddings in seconds instead of minutes
- **üß© Preserves structure**: The exact dictionary is restored, no need to rebuild
- **üìä Maintains efficiency**: All vector data remains optimized for NumPy operations
- **üì± Portability**: Move your processed embeddings between notebooks easily
- **üîÑ Workflow continuity**: Resume your work from exactly where you left off

This approach is especially valuable for large NLP datasets like word embeddings, allowing you to focus on analysis and model building rather than repetitive data processing.

````python
def save_embeddings(embeddings_dict, filename="embeddings_dict.pkl", directory="."):
    """
    Serializes (saves) a dictionary of embeddings to a pickle file.

    Args:
        embeddings_dict (dict): The dictionary of word embeddings.
        filename (str, optional): The name of the pickle file.
            Defaults to "embeddings_dict.pkl".
        directory (str, optional): The directory to save the file in.
            Defaults to the current directory (".").
    """
    filepath = os.path.join(directory, filename)  # Create the full file path

    try:
        # Create the directory if it doesn't exist
        os.makedirs(directory, exist_ok=True)

        with open(filepath, 'wb') as f:
            pickle.dump(embeddings_dict, f)
        print(f"Embeddings saved to: {filepath}")

    except Exception as e:
        print(f"Error saving embeddings: {e}")


def load_embeddings(filename="embeddings_dict.pkl", directory="."):
    """
    Deserializes (loads) a dictionary of embeddings from a pickle file.

    Args:
        filename (str, optional): The name of the pickle file.
            Defaults to "embeddings_dict.pkl".
        directory (str, optional): The directory where the file is located.
            Defaults to the current directory (".").

    Returns:
        dict: The dictionary of word embeddings, or None if an error occurred.
    """
    filepath = os.path.join(directory, filename)

    try:
        with open(filepath, 'rb') as f:
            embeddings_dict = pickle.load(f)
        print(f"Embeddings loaded from: {filepath}")
        return embeddings_dict

    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")
        return None  # Return None to indicate failure
    except Exception as e:
        print(f"Error loading embeddings: {e}")
        return None
````

## Once the function is defined, we‚Äôll call it to save the state of our dictionary üìã into a pickle file, neatly stored in the root directory. This keeps everything organized and ready to roll! ü•íüíæ Let‚Äôs make it happen!

````python
# Save embeddings to a pickle file
save_embeddings(embeddings_dict)
````

This makes them super handy for tasks like finding word similarities, solving analogies, or diving into semantic analysis. In short, these embeddings pack a ton of meaning into a compact form, helping us explore how words connect in a high-dimensional world!


````python
def show_n_first_words(path, n_words=5, show_stats=True, preview_dims=5):
    """
    Display a preview of word vectors from a GloVe embeddings file with
    nicely formatted output for better readability.

    Args:
        path (str): Path to the GloVe embedding file
        n_words (int, optional): Number of words to display. Defaults to 5.
        show_stats (bool, optional): Whether to show statistics about values. Defaults to True.
        preview_dims (int, optional): Number of vector dimensions to show. Defaults to 5.

    Returns:
        bool: True if successful, False if an error occurred
    """
    # Create a visually appealing header
    print("\n" + "‚ïê" * 80)
    print(f"üîç GloVe Embeddings Preview: First {n_words} Words".center(80))
    print("‚ïê" * 80)

    try:
        # Track statistics if requested
        stats = {
            "min_value": float('inf'),
            "max_value": float('-inf'),
            "avg_magnitudes": [],
            "dimensions": set()
        }

        with open(path, 'r', encoding='utf-8') as f:
            # Counter for actual words shown (skipping empty lines)
            words_shown = 0

            for i, line in enumerate(f):
                # Skip empty lines
                values = line.strip().split()
                if not values:
                    continue

                # Extract word and vector
                word = values[0]
                embedding = np.array([float(x) for x in values[1:]])
                embedding_dim = len(embedding)

                # Track statistics if requested
                if show_stats:
                    stats["dimensions"].add(embedding_dim)
                    stats["min_value"] = min(stats["min_value"], embedding.min())
                    stats["max_value"] = max(stats["max_value"], embedding.max())
                    stats["avg_magnitudes"].append(np.mean(np.abs(embedding)))

                # Format the preview of vector values
                preview = ", ".join([f"{x:.4f}" for x in embedding[:preview_dims]])

                # Print word with nice formatting
                print(f"üìå Word #{words_shown+1}: '{word}'")
                print(f"   ‚Ä¢ Dimensions: {embedding_dim}")
                print(f"   ‚Ä¢ Vector preview: [{preview}, ...]")

                # Calculate and show magnitude (L2 norm)
                magnitude = np.linalg.norm(embedding)
                print(f"   ‚Ä¢ Vector magnitude: {magnitude:.4f}")

                # Add separator between words
                print("‚îÄ" * 80)

                # Increment counter and check if we've shown enough words
                words_shown += 1
                if words_shown >= n_words:
                    break

            # Show overall statistics if requested
            if show_stats and words_shown > 0:
                print("\nüìä Overall Statistics (based on previewed words):")
                print(f"   ‚Ä¢ Dimensions observed: {list(stats['dimensions'])}")
                print(f"   ‚Ä¢ Value range: [{stats['min_value']:.4f} to {stats['max_value']:.4f}]")
                if stats["avg_magnitudes"]:
                    avg_mag = sum(stats["avg_magnitudes"]) / len(stats["avg_magnitudes"])
                    print(f"   ‚Ä¢ Average absolute value: {avg_mag:.4f}")
                print("‚ïê" * 80)

            # No words found case
            if words_shown == 0:
                print("‚ö†Ô∏è No valid word vectors found in the file!")

        return True

    except FileNotFoundError:
        print(f"‚ùå Error: File not found at '{path}'")
        print("   ‚Ä¢ Check if the path is correct")
        print("   ‚Ä¢ Ensure the GloVe file has been downloaded")
        print("‚ïê" * 80)
        return False

    except UnicodeDecodeError:
        print(f"‚ùå Error: Cannot decode file at '{path}'")
        print("   ‚Ä¢ The file might be binary or use a different encoding")
        print("   ‚Ä¢ Try using a different encoding parameter")
        print("‚ïê" * 80)
        return False

    except Exception as e:
        print(f"‚ùå Error: {type(e).__name__}: {str(e)}")
        print("‚ïê" * 80)
        return False

# Example usage:
# Show 5 words with additional statistics
# show_n_first_words(PATH, n_words=5, show_stats=True)
````

# Plot some embeddings


The visualize_words function brings word embeddings to life with a 2D scatter plot, powered by dimensionality reduction (PCA or t-SNE). ‚ú® When you run it, here‚Äôs what you‚Äôll see in the visualization:

````python
def visualize_words(words2show, embeddings_index, method='PCA', figsize=(18, 16), show_distances=True):
    """
    Creates a more readable word embeddings visualization with better distribution
    and improved spacing between labels.
    """
    # Similar setup as before, but with distribution improvements
    if not embeddings_index:
        print("‚ùå Error: Embeddings dictionary is empty or not provided")
        return None, None, None

    valid_words = [word for word in words2show if word in embeddings_index]
    if not valid_words:
        print("‚ùå Error: None of the requested words are in the embeddings dictionary")
        return None, None, None

    # Define categories for semantic grouping
    categories = {
        'food': ['burger', 'tortilla', 'bread', 'pizza', 'beef', 'steak', 'fries', 'chips'],
        'drinks': ['water', 'beer', 'tequila', 'wine', 'whisky', 'brandy', 'vodka', 'coffee', 'tea'],
        'countries': ['argentina', 'mexico', 'spain', 'usa', 'france', 'italy', 'greece', 'china'],
        'fruits': ['apple', 'banana', 'orange', 'lemon', 'grapefruit', 'grape', 'strawberry', 'raspberry'],
        'education': ['school', 'work', 'university', 'highschool']
    }

    # Set up distinct, high-contrast colors for better readability
    category_colors = {
        'food': '#FF7F00',       # Bright orange
        'drinks': '#377EB8',     # Blue
        'countries': '#4DAF4A',  # Green
        'fruits': '#FFFF33',     # Yellow
        'education': '#F781BF',  # Pink
        'other': '#999999'       # Gray
    }

    # Map words to categories
    word_to_category = {}
    for word in valid_words:
        found = False
        for category, category_words in categories.items():
            if word.lower() in category_words:
                word_to_category[word] = category
                found = True
                break
        if not found:
            word_to_category[word] = 'other'

    # Extract embeddings
    embeddings = np.array([embeddings_index[word] for word in valid_words])

    # Perform dimensionality reduction
    if method.upper() == 'PCA':
        from sklearn.decomposition import PCA

        # Perform PCA
        reducer = PCA(n_components=2)
        embeddings_2d = reducer.fit_transform(embeddings)

        # Calculate explained variance
        explained_variance = reducer.explained_variance_ratio_
        x_label = f"Principal Component 1 ({explained_variance[0]:.2%} variance)"
        y_label = f"Principal Component 2 ({explained_variance[1]:.2%} variance)"
        title = f'Word Embeddings Visualization: PCA\nTotal variance explained: {sum(explained_variance):.2%}'

    elif method.upper() == 'TSNE':
        from sklearn.manifold import TSNE
        import warnings

        # Suppress specific FutureWarning
        warnings.filterwarnings('ignore', category=FutureWarning, module='sklearn.manifold._t_sne')

        # Dynamically calculate perplexity
        perplexity = min(30, max(5, (len(embeddings) - 1) // 3))

        # Configure t-SNE with updated parameters
        reducer = TSNE(
            n_components=2,
            perplexity=perplexity,
            learning_rate='auto',
            max_iter=1000,  # Updated from n_iter to max_iter
            random_state=42,
            init='pca',
            n_iter_without_progress=300,
            early_exaggeration=12.0,
            verbose=0  # Set to 0 to suppress output
        )

        # Capture any suppressed output if needed
        import io
        import sys

        # Redirect stdout to capture verbose output
        stdout = sys.stdout
        output_capture = io.StringIO()
        sys.stdout = output_capture

        try:
            # Fit and transform the embeddings
            embeddings_2d = reducer.fit_transform(embeddings)
        finally:
            # Restore stdout
            sys.stdout = stdout

        # Optionally, you can access the captured output if needed
        # captured_output = output_capture.getvalue()

        # Update labels and title
        x_label = "t-SNE Dimension 1"
        y_label = "t-SNE Dimension 2"
        title = f'Word Embeddings Visualization: t-SNE\nPerplexity: {perplexity}'

    else:
        raise ValueError(f"Unsupported method: {method}. Choose 'PCA' or 'TSNE'.")


    # ======================================================================
    # Create improved visualization with better distribution
    # ======================================================================
    plt.style.use('seaborn-v0_8-whitegrid')

    # Create a larger figure for better spacing
    fig, ax = plt.subplots(figsize=figsize, dpi=100, facecolor='white')

    # Get colors for each point
    point_colors = [category_colors[word_to_category[word]] for word in valid_words]

    # Add origin lines and subtle quadrant shading for reference
    ax.axhline(y=0, color='gray', linestyle='-', alpha=0.3, zorder=2)
    ax.axvline(x=0, color='gray', linestyle='-', alpha=0.3, zorder=2)

    # Plot points with larger size
    scatter = ax.scatter(
        embeddings_2d[:, 0],
        embeddings_2d[:, 1],
        c=point_colors,
        alpha=0.8,
        s=150,                    # Larger point size
        edgecolors='white',
        linewidth=2,
        zorder=5
    )

    # ======================================================================
    # Improved label placement with adjustText if available
    # ======================================================================
    try:
        from adjustText import adjust_text

        # Create text objects with improved styling
        texts = []
        for i, word in enumerate(valid_words):
            cat_color = category_colors[word_to_category[word]]

            # Make text backgrounds more transparent for better overlap handling
            text = ax.text(
                embeddings_2d[i, 0], embeddings_2d[i, 1], word,
                fontsize=10,  # Smaller font for less crowding
                fontweight='bold',
                color='black',
                bbox=dict(
                    boxstyle="round,pad=0.4",
                    fc=cat_color,
                    ec="gray",
                    alpha=0.7,
                    linewidth=1
                ),
                zorder=10,
                ha='center',
                va='center'
            )
            texts.append(text)

        # Apply adjustText with stronger parameters to prevent overlap
        adjust_text(
            texts,
            force_points=0.5,     # Stronger repulsion from points
            force_text=0.8,       # Stronger text repulsion
            expand_points=(2, 2), # Expanded point radius
            expand_text=(1.5, 1.5), # Expanded text box
            arrowprops=None,
            lim=500               # More iterations for better placement
        )

    except ImportError:
        # Fallback to simpler label placement
        print("For better label placement, install adjustText: pip install adjustText")
        for i, word in enumerate(valid_words):
            cat_color = category_colors[word_to_category[word]]
            ax.annotate(
                word,
                xy=(embeddings_2d[i, 0], embeddings_2d[i, 1]),
                xytext=(10, 10),  # Offset text more from points
                textcoords='offset points',
                fontsize=10,
                fontweight='bold',
                bbox=dict(
                    boxstyle="round,pad=0.4",
                    fc=cat_color,
                    ec="gray",
                    alpha=0.7
                ),
                zorder=10
            )

    # ======================================================================
    # Add cluster annotations with clear positioning
    # ======================================================================
    # Function to find cluster centroids
    def find_cluster_centroid(category_name):
        indices = [i for i, word in enumerate(valid_words)
                  if word_to_category[word] == category_name]
        if indices:
            x = np.mean([embeddings_2d[i, 0] for i in indices])
            y = np.mean([embeddings_2d[i, 1] for i in indices])
            return (x, y)
        return None

    # Add cluster annotations with better positioning
    clusters_to_annotate = {
        'countries': 'Country Cluster',
        'education': 'Education Cluster',
        'food': 'Food Items',
        'fruits': 'Fruits',
        'drinks': 'Beverages'
    }

    for category, label in clusters_to_annotate.items():
        centroid = find_cluster_centroid(category)
        if centroid:
            # Calculate offset to avoid overlapping with points
            x_offset = 1.0 if category in ['countries', 'education'] else -1.0
            y_offset = 1.0 if category in ['education', 'fruits'] else -1.0

            ax.annotate(
                label,
                xy=centroid,
                xytext=(centroid[0] + x_offset, centroid[1] + y_offset),
                fontsize=12,
                fontweight='bold',
                color='#303030',
                backgroundcolor=(1, 1, 1, 0.7),
                arrowprops=dict(
                    arrowstyle='->',
                    color='#505050',
                    lw=1.5,
                    alpha=0.8
                ),
                zorder=15,
                bbox=dict(
                    boxstyle="round,pad=0.4",
                    fc='white',
                    ec="lightgray",
                    alpha=0.8
                )
            )

    # ======================================================================
    # Improve overall chart aesthetics
    # ======================================================================
    # Add grid with less prominence
    ax.grid(True, linestyle='--', alpha=0.3, color='gray', zorder=0)

    # Enhance titles and labels
    ax.set_xlabel(x_label, fontsize=12, labelpad=15)
    ax.set_ylabel(y_label, fontsize=12, labelpad=15)
    ax.set_title(title, fontsize=16, pad=20, fontweight='bold')

    # Add subtle border
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_color('#DDDDDD')
        spine.set_linewidth(1)

    # ======================================================================
    # Add legend with better positioning
    # ======================================================================
    # Create compact legend
    legend_elements = []
    used_categories = sorted(set(word_to_category.values()))

    for category in used_categories:
        if category != 'other':  # Skip 'other' category
            legend_elements.append(
                plt.Line2D(
                    [0], [0],
                    marker='o',
                    color='w',
                    markerfacecolor=category_colors[category],
                    markersize=10,
                    label=category.title()
                )
            )

    # Create legend with horizontal layout and place at bottom
    legend = ax.legend(
        handles=legend_elements,
        title="Word Categories",
        loc='upper center',
        bbox_to_anchor=(0.5, -0.15),  # Changed from -0.05 to -0.15 to move it lower
        ncol=len(legend_elements),
        frameon=True,
        fancybox=True,
        shadow=True
    )

    # Ensure the legend title is bold
    plt.setp(legend.get_title(), fontweight='bold')

    # Ensure proper spacing around the figure
    plt.tight_layout(rect=[0, 0.1, 1, 0.95])

    # Save the visualization with a clear filename
    import time
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    filename = f"distributed_word_embeddings_{method.lower()}_{timestamp}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"üíæ Visualization saved as: {filename}")

    # ======================================================================
    # Step 7: Show distance analysis if requested
    # ======================================================================
    if show_distances and len(valid_words) > 1:
        print("\n" + "‚ïê" * 80)
        print("üìè Word Relationships Analysis".center(80))
        print("‚ïê" * 80)

        # Calculate distance matrix
        distances = squareform(pdist(embeddings_2d))

        # Find closest pairs
        print("\nüîç Closest Word Pairs".center(80))
        print("‚îÄ" * 80)
        print(f"{'Word Pair':^50} {'Distance':^15} {'Categories':^15}")
        print("‚îÄ" * 80)

        # Get indices of closest pairs
        flat_indices = np.argsort(distances, axis=None)
        indices = np.array(np.unravel_index(flat_indices, distances.shape)).T

        # Filter out duplicates and self-pairs
        unique_pairs = []
        seen = set()
        for i, j in indices:
            if i != j and i < j:  # Avoid self-pairs and duplicates
                pair = (i, j)
                if pair not in seen:
                    seen.add(pair)
                    unique_pairs.append(pair)

        # Show top 5 closest pairs
        for i, j in unique_pairs[:5]:
            word1, word2 = valid_words[i], valid_words[j]
            cat1 = word_to_category[word1]
            cat2 = word_to_category[word2]
            categories = f"{cat1}/{cat2}" if cat1 != cat2 else cat1
            print(f"'{word1}' ‚Üî '{word2}'{'':<5} {distances[i, j]:.3f}{'':<10} {categories}")

        # Show farthest pairs
        print("\nüî≠ Most Distant Word Pairs".center(80))
        print("‚îÄ" * 80)
        print(f"{'Word Pair':^50} {'Distance':^15} {'Categories':^15}")
        print("‚îÄ" * 80)

        # Show top 5 farthest pairs
        for i, j in unique_pairs[-5:][::-1]:
            word1, word2 = valid_words[i], valid_words[j]
            cat1 = word_to_category[word1]
            cat2 = word_to_category[word2]
            categories = f"{cat1}/{cat2}" if cat1 != cat2 else cat1
            print(f"'{word1}' ‚Üî '{word2}'{'':<5} {distances[i, j]:.3f}{'':<10} {categories}")

        print("\n" + "‚ïê" * 80)
        print("üí° Note: Categories show the semantic groupings used for coloring".center(80))
        print("‚ïê" * 80)

    # Show the plot
    plt.show()

    # Return objects for further customization
    return fig, ax, embeddings_2d

# Example usage:
# _ = visualize_words(words, embeddings_dict, method='PCA', figsize=(18, 16), show_distances=True)
````

# Word Embedding Analogies: Discovering Semantic Relationships

## Understanding Word Analogies

### The Analogy Concept

Consider the classic analogy:
**`Man üë® is to King ü§¥ as Woman üë© is to ...`**

The expected answer is `Queen` üë∏, demonstrating a deep semantic relationship between words.

### Mathematical Representation of Word Analogies

The analogy is solved using vector arithmetic:

$$A - B + C = D$$

Where:
- A = "Man"
- B = Captures the gender attribute
- C = "Woman"
- D = The target word (in this case, "Queen")

## Vector Similarity Techniques

### Why Cosine Similarity?

When working with word embeddings, multiple similarity metrics exist:
- Euclidean Distance
- Manhattan Distance
- Pearson Correlation
- **Cosine Similarity** (Most Commonly Used)

### Cosine Similarity: The Preferred Method

#### Mathematical Formula

$$\text{cosine similarity} = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \|\mathbf{B}\|}$$

#### Key Characteristics
- Measures angle between vectors
- Independent of vector magnitude
- Ranges from -1 to 1
- 1 indicates identical direction
- 0 indicates orthogonal vectors
- -1 indicates opposite directions

## Computational Approaches

### Possible Implementation Methods
1. NumPy
2. SciPy
3. **PyTorch**
4. TensorFlow
5. Manual Calculation

````python
# Function to calculate cosine similarity
def cosine_similarity(vec1, vec2):
    """Calculate cosine similarity between two vectors"""
    dot_product = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    return dot_product / (norm1 * norm2)

print(f"Cosine similarity between man and woman using custom function: {F.cosine_similarity(example_vector1, example_vector2).item():.4f}")
````

## Understanding the Word Analogy Function

### üéØ Function Purpose
The `solve_analogy()` function is designed to uncover semantic relationships by solving word analogies of the form:

**`A is to B as C is to ?`**

### üß© Analogy Solving Process

#### Step-by-Step Mechanism

1. **Embedding Retrieval** üìä
   - Extract vector representations for three input words
   - Utilizes the pre-defined `embeddings_dictionary`

2. **Target Vector Calculation** üéØ
   - Compute the semantic transformation vector
   - Represents the relationship between the first two words
   - Mathematically expressed as: `target_vector = word2 - word1 + word3`

3. **Similarity Exploration** üîç
   - Compare the target vector with embeddings of all words in the dictionary
   - Calculate cosine similarity for each comparison
   - Track potential analogy matches

4. **Ranking and Selection** üìà
   - Sort similarities to find the most semantically related word
   - Exclude the input words to ensure a unique solution

### üåü Key Computational Steps

```python
def solve_analogy(word1, word2, word3, embeddings_dict):
    # 1. Retrieve word embeddings
    vec1 = embeddings_dict[word1]
    vec2 = embeddings_dict[word2]
    vec3 = embeddings_dict[word3]
    
    # 2. Calculate target vector
    target_vector = vec2 - vec1 + vec3
    
    # 3 & 4. Compute similarities and store results
    similarities = []
    for word, vec in embeddings_dict.items():
        # Avoid using input words
        if word not in [word1, word2, word3]:
            similarity = cosine_similarity(target_vector, vec)
            similarities.append((word, similarity))
    
    # 5 & 6. Sort and return best match
    similarities.sort(key=lambda x: x[1], reverse=True)
    return similarities[0][0]
```

````python
# Analogy function A is B as X is to Y
def analogy(word1, word2, word3, embeddings_dict):
    """
    Find words that complete the analogy: word1 is to word2 as word3 is to ?.
    Based on the vector offset method: word2 - word1 + word3

    Parameters:
    word1, word2, word3: String words in the embeddings dictionary
    embeddings_dict: Dictionary mapping words to their embeddings

    Returns:
    List of (word, score) tuples with the top matching words
    """
    # Check if all words exist in the embeddings dictionary
    for word in [word1, word2, word3]:
        if word not in embeddings_dict:
            print(f"Error: '{word}' not found in embeddings dictionary.")
            return None

    # Get the embeddings
    embedding1 = embeddings_dict[word1]
    embedding2 = embeddings_dict[word2]
    embedding3 = embeddings_dict[word3]

    # Calculate target vector: word2 - word1 + word3
    target_vector = embedding2 - embedding1 + embedding3

    # Find closest words to target vector
    similarities = []
    for word, embedding in embeddings_dict.items():
        # Skip the input words
        if word in [word1, word2, word3]:
            continue

        # Calculate cosine similarity
        similarity = cosine_similarity(target_vector, embedding)
        similarities.append((word, similarity))

    # Sort by similarity (highest first)
    similarities.sort(key=lambda x: x[1], reverse=True)

    # Return top 10 results and best result
    return similarities[:10], similarities[0][0]
````

# Most Similar Function

Now the next challenge was to create a function that returns similar words based on an input. For this, we do seomthing similar than before for the analogy function, except that now instead of the target vector, we calculate cosine similarity of the embedding coefficients of the input word against each one of the words cotained in the dictionary.

````python
# Find Most Similar function
def find_most_similar(word, embeddings_dict, top_n=10):
    """
    Find words most similar to the input word using cosine similarity.

    Parameters:
    word: String word to find similar words for
    embeddings_dict: Dictionary mapping words to their embeddings
    top_n: Number of similar words to return

    Returns:
    List of (word, similarity) tuples with the top_n most similar words
    """
    # Check if word exists in embeddings
    if word not in embeddings_dict:
        print(f"Error: '{word}' not found in embeddings dictionary.")
        return None

    # Get the embedding for the input word
    word_embedding = embeddings_dict[word]

    # Calculate similarity with all other words
    similarities = []
    for other_word, other_embedding in embeddings_dict.items():
        # Skip the input word
        if other_word == word:
            continue

        # Calculate cosine similarity
        similarity = cosine_similarity(word_embedding, other_embedding)
        similarities.append((other_word, similarity))

    # Sort by similarity (highest first)
    similarities.sort(key=lambda x: x[1], reverse=True)

    # Return top N results
    return similarities[:top_n]
````

````python
# Function to clean and process text
def process_shakespeare_text(file_path):
    print("-" * 50)
    print("üîÑ Processing Shakespeare's text üîÑ")
    print("-" * 50)

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()

        # Remove Project Gutenberg header and footer
        start_marker = "THE COMPLETE WORKS OF WILLIAM"
        end_marker = "End of Project Gutenberg's The Complete Works of William Shakespeare"

        # Find the positions
        start_pos = text.find(start_marker)
        end_pos = text.find(end_marker)

        if start_pos != -1 and end_pos != -1:
            text = text[start_pos:end_pos]

        # Remove non-alphabetic characters and convert to lowercase
        text = re.sub(r'[^a-zA-Z\s]', ' ', text)
        text = text.lower()

        # Tokenize into sentences and words
        sentences = []
        stop_words = set(stopwords.words('english'))  # Load stopwords

        for line in text.split('\n'):
            if line.strip():  # Skip empty lines
                tokens = word_tokenize(line)
                if len(tokens) > 2:  # Only include sentences with at least 3 words
                    sentences.append(tokens)

        print(f"‚úÖ Processed {len(sentences)} sentences from Shakespeare's works")
        return sentences

    except Exception as e:
        print(f"‚ùå Error processing Shakespeare's text: {e}")
        return None
````

````python
# Train Word2Vec model on Shakespeare's text
def train_word2vec(sentences, vector_size=100, window=3, min_count=1):
    print("-" * 50)
    print(f"üß† Training Word2Vec model (dim={vector_size}) üß†")
    print("-" * 50)

    model = Word2Vec(sentences=sentences,
                    vector_size=vector_size,
                    window=window,
                    min_count=min_count,
                    workers=4)

    model_path = os.path.join('shakespeare_data', f'shakespeare_word2vec_{vector_size}d.model')
    model.save(model_path)

    print(f"‚úÖ Word2Vec model trained and saved to {model_path}")
    print(f"Vocabulary size: {len(model.wv.key_to_index)}")

    return model
````

````python
# Create embeddings dictionary from Word2Vec model
def create_embeddings_dict(model):
    embeddings_dict = {}
    for word in model.wv.key_to_index:
        embeddings_dict[word] = model.wv[word]

    return embeddings_dict
````

````python
# Visualize word embeddings
def visualize_words(words2show, embeddings_dict, method='PCA', figsize=(12, 10), fontsize=10):
    """
    Visualize word embeddings using dimensionality reduction techniques.
    """
    # Extract embeddings for the words to show
    valid_words = [word for word in words2show if word in embeddings_dict]
    if not valid_words:
        print("None of the words are present in the embeddings dictionary.")
        return

    embeddings = np.array([embeddings_dict[word] for word in valid_words])

    # Select method of dimensional reduction
    if method == 'PCA':
        reducer = PCA(n_components=2)
        embeddings_2d = reducer.fit_transform(embeddings)
        explained_variance = reducer.explained_variance_ratio_
        x_label = f"Principal Component 1 ({explained_variance[0]:.2%} variance explained)"
        y_label = f"Principal Component 2 ({explained_variance[1]:.2%} variance explained)"
        title = f'Shakespeare Word Embeddings (PCA)\nTotal variance explained: {sum(explained_variance):.2%}'

    elif method == 'TSNE':
        # First reduce dimensionality with PCA to speed up t-SNE
        pca = PCA(n_components=min(50, len(embeddings)))
        embeddings_pca = pca.fit_transform(embeddings)
        reducer = TSNE(n_components=2, random_state=42, perplexity=min(30, len(embeddings)-1))
        embeddings_2d = reducer.fit_transform(embeddings_pca)
        x_label = "t-SNE dimension 1"
        y_label = "t-SNE dimension 2"
        title = f'Shakespeare Word Embeddings (t-SNE)'

    else:
        print(f"Unknown method: {method}")
        return

    # Create the figure
    fig, ax = plt.subplots(figsize=figsize)

    # Create scatter plot
    ax.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.8, s=80)

    # Add annotations for words
    for i, word in enumerate(valid_words):
        ax.annotate(word, xy=(embeddings_2d[i, 0], embeddings_2d[i, 1]),
                   xytext=(5, 2), textcoords='offset points',
                   fontsize=fontsize)

    # Set labels and title
    ax.set_xlabel(x_label, fontsize=18)
    ax.set_ylabel(y_label, fontsize=18)
    ax.set_title(title, fontsize=18, pad=20)

    # Save visualization
    plt.tight_layout()
    plt.savefig(f'shakespeare_embeddings_{method}.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Print statistics
    print(f"Successfully visualized {len(valid_words)} words from Shakespeare")

    if len(valid_words) < len(words2show):
        missing = set(words2show) - set(valid_words)
        print(f"Could not find embeddings for: {missing}")
````
