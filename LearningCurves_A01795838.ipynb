# Las Curvas de Aprendizaje en Modelos de Aprendizaje Autom치tico

Las Curvas de Aprendizaje son una herramienta fundamental para evaluar el rendimiento y comportamiento de modelos de aprendizaje autom치tico.

## Definici칩n:
Las curvas de aprendizaje muestran el efecto de agregar m치s muestras durante el proceso de entrenamiento, evaluando el rendimiento estad칤stico del modelo en t칠rminos de puntuaci칩n de entrenamiento y puntuaci칩n de prueba. Este m칠todo permite visualizar y diagnosticar problemas de sesgo y varianza en los modelos.

## Caracter칤sticas Principales:

### Comportamiento del Clasificador Naive Bayes
- La puntuaci칩n de entrenamiento es muy alta con pocas muestras y disminuye al aumentar el n칰mero de muestras.
- La puntuaci칩n de prueba comienza muy baja y aumenta con m치s muestras.
- Las puntuaciones de entrenamiento y prueba convergen cuando se utilizan todas las muestras.

### Comportamiento del Clasificador SVM
- La puntuaci칩n de entrenamiento se mantiene alta independientemente del tama침o del conjunto de entrenamiento.
- La puntuaci칩n de prueba aumenta con el tama침o del conjunto de datos hasta alcanzar una meseta.
- La presencia de esta meseta indica que agregar m치s datos no mejorar치 significativamente el rendimiento del modelo.

## Diagn칩stico de Problemas:

### Subajuste (Underfitting)
- Se identifica cuando tanto las puntuaciones de entrenamiento como de prueba son bajas.
- Indica que el modelo es demasiado simple para capturar los patrones en los datos.

### Sobreajuste (Overfitting)
- Se observa cuando hay una gran brecha entre las puntuaciones de entrenamiento y prueba.
- La puntuaci칩n de entrenamiento es significativamente m치s alta que la de prueba.
- Sugiere que el modelo ha memorizado los datos de entrenamiento pero no generaliza bien.

## Consideraciones de Escalabilidad

La complejidad computacional var칤a seg칰n el algoritmo:
- El clasificador SVM muestra un aumento r치pido en el tiempo de ajuste y puntuaci칩n con m치s muestras.
- El clasificador Naive Bayes escala mejor y muestra menor complejidad en tiempos de ajuste y puntuaci칩n.

```python
# Aplicaremos lo m칤nimo de transformaciones, en particular para datos perdidos:
# Creamos un pipeline para variables num칠ricas que reemplaza valores faltantes con la mediana
num_pipe = Pipeline(steps = [('numImp', SimpleImputer(strategy='median')),
                             ])
num_pipe_nombres = [0]   # ['Page total likes'] # Solo se aplica a la columna 'Page total likes'


# Las ordinales las identificamos como tales:
# Pipeline para variables ordinales (que tienen un orden natural)
ord_pipe = Pipeline(steps =  [('OrdImp', SimpleImputer(strategy='most_frequent')), # Primero rellena valores faltantes con el valor m치s frecuente
                              ('Ordfun', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1 )) # Luego codifica las variables ordinales
                              ])
ord_pipe_nombres = [3,4,5]   # ['Post Month', 'Post Weekday', 'Post Hour'] # Se aplica a estas tres columnas que tienen orden natural

# Las nominales las transformamos con One-Hot-Encoder:
# Pipeline para variables categ칩ricas sin orden (nominales)
nom_pipe = Pipeline(steps = [('ohe', OneHotEncoder(drop='first', # Elimina la primera categor칤a para evitar multicolinealidad
                                                   handle_unknown='ignore', # Ignora categor칤as nuevas en datos de prueba
                                                   ))])
nom_pipe_nombres = [1,2,6]   # ['Type', 'Category', 'Paid'] # Se aplica a estas variables categ칩ricas

# Conjuntamos las transformaciones que aplicaremos:
# Combina todos los pipelines anteriores en un solo transformador
columnasTransformer = ColumnTransformer(transformers = [('numpow', num_pipe, num_pipe_nombres), # Pipeline num칠rico
                                                        ('catOrd', ord_pipe, ord_pipe_nombres),  # Pipeline ordinal
                                                        ('catNom', nom_pipe, nom_pipe_nombres)   # Pipeline nominal
                                                        ],
                                        remainder='passthrough') # Mantiene cualquier columna no especificada sin cambios
```

```python
# Solo para saber la nueva cantidad de columnas despu칠s de las transformaciones:

# Aplicamos las transformaciones definidas en columnasTransformer al conjunto de entrenamiento
# fit_transform() realiza dos operaciones:
# 1. fit: Aprende los par치metros de la transformaci칩n de los datos de entrenamiento
# 2. transform: Aplica la transformaci칩n usando los par치metros aprendidos
XtrainT = columnasTransformer.fit_transform(Xtrain)  # Ajustamos con Train...

# Imprimimos las dimensiones antes y despu칠s de la transformaci칩n
print('Variables de entrada original:', Xtrain.shape)    # Muestra (n_muestras, 7) - 7 variables originales
print('Variables de entrada transformadas:', XtrainT.shape)  # Muestra el nuevo n칰mero de variables despu칠s de:
                                                            # - One-hot encoding para variables categ칩ricas
                                                            # - Codificaci칩n ordinal para variables ordinales
                                                            # - Imputaci칩n para valores faltantes
```

```python
# El valor de RMSE-Root-Mean-Square-Error, se utiliza para medir el desempe침o
# m칤nimo que debiera obtener un modelo de regresi칩n, es decir, este valor
# se toma usualmente como el valor del modelo subentrenado y se compara generalmente
# con el valor de la desviaci칩n est치ndar de la variable de salida, ya sea que la
# est칠s tomando de manera original o con los datos transformados.
# Veamos el valor de ambas:

# Calculamos y mostramos las desviaciones est치ndar de los datos originales y transformados
print('Desv-Est y-original: %.3f' % ytrain.values.std(ddof=1))      # Usemos la desviaci칩n est치ndar muestral ajustando.
print('Desv-Est y-logaritmo: %.3f' % ytrainlog.values.std(ddof=1))  # los grados de libertad.
```
