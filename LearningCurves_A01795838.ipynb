# Las Curvas de Aprendizaje en Modelos de Aprendizaje Automático

Las Curvas de Aprendizaje son una herramienta fundamental para evaluar el rendimiento y comportamiento de modelos de aprendizaje automático.

## Definición:
Las curvas de aprendizaje muestran el efecto de agregar más muestras durante el proceso de entrenamiento, evaluando el rendimiento estadístico del modelo en términos de puntuación de entrenamiento y puntuación de prueba. Este método permite visualizar y diagnosticar problemas de sesgo y varianza en los modelos.

## Características Principales:

### Comportamiento del Clasificador Naive Bayes
- La puntuación de entrenamiento es muy alta con pocas muestras y disminuye al aumentar el número de muestras.
- La puntuación de prueba comienza muy baja y aumenta con más muestras.
- Las puntuaciones de entrenamiento y prueba convergen cuando se utilizan todas las muestras.

### Comportamiento del Clasificador SVM
- La puntuación de entrenamiento se mantiene alta independientemente del tamaño del conjunto de entrenamiento.
- La puntuación de prueba aumenta con el tamaño del conjunto de datos hasta alcanzar una meseta.
- La presencia de esta meseta indica que agregar más datos no mejorará significativamente el rendimiento del modelo.

## Diagnóstico de Problemas:

### Subajuste (Underfitting)
- Se identifica cuando tanto las puntuaciones de entrenamiento como de prueba son bajas.
- Indica que el modelo es demasiado simple para capturar los patrones en los datos.

### Sobreajuste (Overfitting)
- Se observa cuando hay una gran brecha entre las puntuaciones de entrenamiento y prueba.
- La puntuación de entrenamiento es significativamente más alta que la de prueba.
- Sugiere que el modelo ha memorizado los datos de entrenamiento pero no generaliza bien.

## Consideraciones de Escalabilidad

La complejidad computacional varía según el algoritmo:
- El clasificador SVM muestra un aumento rápido en el tiempo de ajuste y puntuación con más muestras.
- El clasificador Naive Bayes escala mejor y muestra menor complejidad en tiempos de ajuste y puntuación.

```python
# Aplicaremos lo mínimo de transformaciones, en particular para datos perdidos:
# Creamos un pipeline para variables numéricas que reemplaza valores faltantes con la mediana
num_pipe = Pipeline(steps = [('numImp', SimpleImputer(strategy='median')),
                             ])
num_pipe_nombres = [0]   # ['Page total likes'] # Solo se aplica a la columna 'Page total likes'


# Las ordinales las identificamos como tales:
# Pipeline para variables ordinales (que tienen un orden natural)
ord_pipe = Pipeline(steps =  [('OrdImp', SimpleImputer(strategy='most_frequent')), # Primero rellena valores faltantes con el valor más frecuente
                              ('Ordfun', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1 )) # Luego codifica las variables ordinales
                              ])
ord_pipe_nombres = [3,4,5]   # ['Post Month', 'Post Weekday', 'Post Hour'] # Se aplica a estas tres columnas que tienen orden natural

# Las nominales las transformamos con One-Hot-Encoder:
# Pipeline para variables categóricas sin orden (nominales)
nom_pipe = Pipeline(steps = [('ohe', OneHotEncoder(drop='first', # Elimina la primera categoría para evitar multicolinealidad
                                                   handle_unknown='ignore', # Ignora categorías nuevas en datos de prueba
                                                   ))])
nom_pipe_nombres = [1,2,6]   # ['Type', 'Category', 'Paid'] # Se aplica a estas variables categóricas

# Conjuntamos las transformaciones que aplicaremos:
# Combina todos los pipelines anteriores en un solo transformador
columnasTransformer = ColumnTransformer(transformers = [('numpow', num_pipe, num_pipe_nombres), # Pipeline numérico
                                                        ('catOrd', ord_pipe, ord_pipe_nombres),  # Pipeline ordinal
                                                        ('catNom', nom_pipe, nom_pipe_nombres)   # Pipeline nominal
                                                        ],
                                        remainder='passthrough') # Mantiene cualquier columna no especificada sin cambios
```

```python
# Solo para saber la nueva cantidad de columnas después de las transformaciones:

# Aplicamos las transformaciones definidas en columnasTransformer al conjunto de entrenamiento
# fit_transform() realiza dos operaciones:
# 1. fit: Aprende los parámetros de la transformación de los datos de entrenamiento
# 2. transform: Aplica la transformación usando los parámetros aprendidos
XtrainT = columnasTransformer.fit_transform(Xtrain)  # Ajustamos con Train...

# Imprimimos las dimensiones antes y después de la transformación
print('Variables de entrada original:', Xtrain.shape)    # Muestra (n_muestras, 7) - 7 variables originales
print('Variables de entrada transformadas:', XtrainT.shape)  # Muestra el nuevo número de variables después de:
                                                            # - One-hot encoding para variables categóricas
                                                            # - Codificación ordinal para variables ordinales
                                                            # - Imputación para valores faltantes
```

```python
# El valor de RMSE-Root-Mean-Square-Error, se utiliza para medir el desempeño
# mínimo que debiera obtener un modelo de regresión, es decir, este valor
# se toma usualmente como el valor del modelo subentrenado y se compara generalmente
# con el valor de la desviación estándar de la variable de salida, ya sea que la
# estés tomando de manera original o con los datos transformados.
# Veamos el valor de ambas:

# Calculamos y mostramos las desviaciones estándar de los datos originales y transformados
print('Desv-Est y-original: %.3f' % ytrain.values.std(ddof=1))      # Usemos la desviación estándar muestral ajustando.
print('Desv-Est y-logaritmo: %.3f' % ytrainlog.values.std(ddof=1))  # los grados de libertad.
```

#### **NOTA:**
#### **Estricamente, desde el punto de vista matemático las desviaciones estándar que uno calcula deben ser las muestrales, es decir, std(ddof=1). Sin embargo, en la práctica es muy común usar la función con su valor predeterminado que es ddof=0 y que se traduce a calcular la desviación estándar poblacional. Repito, en la práctica suele pasarse por alto este ajuste porque cuando se tienen datos de miles o más regitros, la diferencia entre el resultado problacional y el muestral es mínima. En esta actividad lo pongo solamente para recordarlo y en dado caso cuando tengas menos de 100 registros sí hacer este ajuste, porque la diferencia entre ambos ya podría ser importante.**

```python
# Como se va a utilizar Validación-Cruzada, concatenamos los conjuntos de entrenamiento y validación
# en uno nuevo conjunto aumentado que llamaremos trainval para utilizar como entrenamiento:

Xtrainval = np.concatenate((Xtrain, Xval), axis=0)  # Concatenamos las características de Xtrain y Xval a lo largo de las filas
ytrainvallog = np.concatenate((ytrainlog, yvallog), axis=0)  # Concatenamos las etiquetas transformadas de ytrainlog y yvallog a lo largo de las filas
```

## **Ejercicio 3**


```python
# Al ejecutar las siguientes líneas de código determina si el modelo de
# Bosque Aleatorio con sus valores de hiperparámetros predeterminadas está
# Subentrenado o Sobreentrenado. De ser así, busca los valores de sus
# hiperparámetros que consideres más adecuados para que ya no
# esté sub-o-sobreentrenado:


# ++++++++++++++ Incluye aquí tus ajustes ++++++++++++++++++++++++++++++

#### HIPERPARAMETOS
####### Creamos el modelo Random Forest con los siguientes parámetros:
####
elmodelo_RF = RandomForestRegressor(
    n_estimators=200,        # Número de árboles en el bosque - más árboles = más robusto
    max_depth=3,             # Profundidad máxima de cada árbol - limita el sobreajuste
    min_samples_split=10,    # Mínimo de muestras necesarias para dividir un nodo
                             # Valor alto ayuda a prevenir el sobreajuste
    min_samples_leaf=5,      # Mínimo de muestras requeridas en un nodo hoja
                             # Valor alto genera árboles más balanceados
    max_features='sqrt',     # Número de features a considerar en cada split
                             # sqrt = raíz cuadrada del total de features
    bootstrap=True,          # Usar o no bootstrap samples para construir árboles
                             # True = cada árbol usa una muestra aleatoria con reemplazo
    random_state=42          # Semilla para reproducibilidad de resultados
)

# +++++++++++++ Termina sección de realizar ajustes +++++++++++++++++++++



mipipe = Pipeline(steps=[('ct',columnasTransformer),('m', elmodelo_RF)])

train_sizes, train_scores, val_scores = learning_curve(estimator=mipipe,
                                                        X=Xtrainval,
                                                        y=np.ravel(ytrainvallog),
                                                        cv=5,
                                                        train_sizes= np.linspace(0.1, 1.0, 10),
                                                        scoring= 'neg_mean_squared_error',  # error MSE
                                                        n_jobs=-1)



# Calculamos los promedios y desviación estándar de entrenamiento para RMSE,
# como tenemos los de MSE, debemos obtener su raíz cuadrada. El negativo es porque
# sklearn nos devuelve en realidad el negativo del Error Cuadrático Medio MSE:

rmse_train_scores = np.sqrt(-train_scores)
rmse_val_scores = np.sqrt(-val_scores)

train_mean = np.mean(rmse_train_scores, axis=1)
train_std = np.std(rmse_train_scores, axis=1, ddof=1)   # aquí también hacemos el ajuste de la desviación estándar muestral.
val_mean = np.mean(rmse_val_scores, axis=1)
val_std = np.std(rmse_val_scores, axis=1, ddof=1)


plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training RMSE')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')

plt.plot(train_sizes, val_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation RMSE')
plt.fill_between(train_sizes, val_mean + val_std, val_mean - val_std, alpha=0.15, color='green')

plt.title('Errors : Learning Curve')
plt.xlabel('Training Data Size')
plt.ylabel('Modelo Error RMSE')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# Modelo Inicial con Hiperparámetros Predeterminados

```python
# Preparación de los datos
X = data.iloc[:, 0:7]
y = data[['Lifetime Post Consumers']]

# Mapear la variable categórica 'Type' a valores numéricos
X['Type'] = X['Type'].map({'Photo': 1, 'Status': 2, 'Link': 3, 'Video': 4})

# División de los datos en conjuntos de entrenamiento, validación y prueba
Xtrain, Xtv, ytrain, ytv = train_test_split(X, y, train_size=0.6, shuffle=True, random_state=5)
Xval, Xtest, yval, ytest = train_test_split(Xtv, ytv, test_size=0.5, shuffle=True, random_state=7)

# Aplicar transformación logarítmica a las variables objetivo
ytrainlog = np.log(ytrain)
yvallog = np.log(yval)
ytestlog = np.log(ytest)

# Concatenar conjuntos de entrenamiento y validación
Xtrainval = pd.concat([Xtrain, Xval])
ytrainvallog = pd.concat([ytrainlog, yvallog])

# Definir los nombres de las columnas para cada tipo de transformación
num_pipe_nombres = [0]  # ['Page total likes']
ord_pipe_nombres = [3, 4, 5]  # ['Post Month', 'Post Weekday', 'Post Hour']
nom_pipe_nombres = [1, 2, 6]  # ['Type', 'Category', 'Paid']

# Crear los pipelines para cada tipo de dato
num_pipe = Pipeline(steps=[
    ('numImp', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

ord_pipe = Pipeline(steps=[
    ('OrdImp', SimpleImputer(strategy='most_frequent')),
    ('Ordfun', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)),
    ('scaler', StandardScaler())
])

nom_pipe = Pipeline(steps=[
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore')),
])

# Combinar las transformaciones en un ColumnTransformer
columnasTransformer = ColumnTransformer(transformers=[
    ('numpow', num_pipe, num_pipe_nombres),
    ('catOrd', ord_pipe, ord_pipe_nombres),
    ('catNom', nom_pipe, nom_pipe_nombres),
], remainder='passthrough')



# Crear el modelo MLP con hiperparámetros predeterminados
elmodelo_MLP = MLPRegressor(
    hidden_layer_sizes=(1, 10, 30),  # Tres capas ocultas
    activation='relu',
    solver='adam',
    alpha=0.00012,
    learning_rate_init=0.02,
    max_iter=500,
    random_state=42,
    momentum=0.9,
    early_stopping=False
)

# Crear el pipeline
mipipe = Pipeline(steps=[('ct', columnasTransformer), ('m', elmodelo_MLP)])

# Calcular la curva de aprendizaje
train_sizes, train_scores, val_scores = learning_curve(
    estimator=mipipe,
    X=Xtrainval,
    y=np.ravel(ytrainvallog),
    cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='neg_mean_squared_error',
    n_jobs=-1
)

# Calcular RMSE
rmse_train_scores = np.sqrt(-train_scores)
rmse_val_scores = np.sqrt(-val_scores)

train_mean = np.mean(rmse_train_scores, axis=1)
train_std = np.std(rmse_train_scores, axis=1, ddof=1)
val_mean = np.mean(rmse_val_scores, axis=1)
val_std = np.std(rmse_val_scores, axis=1, ddof=1)

# Graficar la curva de aprendizaje con valores de RMSE
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training RMSE')
for x, y in zip(train_sizes, train_mean):
    plt.text(x, y, f"{y:.2f}", color='blue', fontsize=8, ha='center', va='bottom')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')

plt.plot(train_sizes, val_mean, color='green', marker='s', markersize=5, linestyle='--', label='Validation RMSE')
for x, y in zip(train_sizes, val_mean):
    plt.text(x, y, f"{y:.2f}", color='green', fontsize=8, ha='center', va='bottom')
plt.fill_between(train_sizes, val_mean + val_std, val_mean - val_std, alpha=0.15, color='green')

plt.title('Curva de Aprendizaje - MLP (Hiperparámetros Predeterminados)')
plt.xlabel('Tamaño del Conjunto de Entrenamiento')
plt.ylabel('Error RMSE')
plt.grid()
plt.legend(loc='best')
plt.show()
```
